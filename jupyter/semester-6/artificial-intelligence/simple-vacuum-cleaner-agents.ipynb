```python
# Q1.a

import random

class Environment():
    def __init__(self):
        # Let's now instantiate Rooms with 'Clean' or 'Dirty' randomly
        self.locationCondition = {'A':random.choice(['Clean','Dirty']), 'B':random.choice(['Clean','Dirty'])}
        

class TableDrivenVacuumAgent():
    def __init__(self, Environment):
        print(Environment.locationCondition)
        # Instantiate performace measurement
        self.Environment = Environment
        self.Score = 0
        self.vacuumLocation = random.choice(['A','B'])   # In start it will be current Location and when moved right or left this variable acts as previous Location
        self.LocationCondition = self.Environment.locationCondition[self.vacuumLocation]     # In start it will be current Location condition and when moved right or left this variable acts as previous Location condition
        self.Table = { ('A','Clean'): 'right',
                       ('A','Dirty'): 'suck',
                       ('B','Clean'): 'left',
                       ('B','Dirty'): 'suck',
                       ('A','Clean', 'B','Dirty'): 'suck',
                       ('A','Dirty', 'B','Dirty'): 'suck',
                       ('B','Clean', 'A','Dirty'): 'suck',
                       ('B','Dirty', 'A','Dirty'): 'suck'
                                                           }
    
    def _change_location(self, location):   # This function helps in selecting next Location 
        if location == 'A':
            return 'B'
        return 'A'
        
    def _map_action_to_actuator(self, action, location):
        if action == 'right' or action == 'left':
            print('Moving to Location {}...'.format( self._change_location(location) ))            
            self.Score -= 1
        
        elif action == 'suck':
            # Below condtion only checks a bug which I faced i.e If previous and current Location both were Dirty it was not printing Moving to ... that's why and also handling score
            if self.LocationCondition == 'Dirty' and location != self.vacuumLocation:
                self.Score -= 1
                print('Moving to Location {}...'.format( self._change_location(location) ))  # Should move right or left
            
            print('Location {} is Dirty.\nLocation {} has been Cleaned.'.format( location, location ))            
            self.Environment.locationCondition[location] = 'Clean'
            self.Score += 1
        return
            
        
    def start(self):
        print('Vacuum randomly placed at Location', self.vacuumLocation)
        action = self.Table[(self.vacuumLocation, self.LocationCondition)]
        self._map_action_to_actuator(action, self.vacuumLocation)
        
        nextLocation = self._change_location(self.vacuumLocation)  # Now as I have to move to next Location
        nextLocationCondition = self.Environment.locationCondition[nextLocation]
        self.proceed(nextLocation, nextLocationCondition)  
        return
    
    # proceed handles states like (A,clean, B,clean) etc where Vacuum has moved to next Location
    def proceed(self, nextLocation, nextLocationCondition):
        # if next Location is clean don't do anything
        if nextLocationCondition == 'Dirty':
            action = self.Table[(self.vacuumLocation, self.LocationCondition, nextLocation, nextLocationCondition)]
            self._map_action_to_actuator(action, nextLocation)
        
        print(self.Environment.locationCondition)
        print('Performance Measurement: ', self.Score, '\n')
        return
```

```python
for i in range(0,100):
    theEnvironment = Environment()
    theAgent = TableDrivenVacuumAgent(theEnvironment)
    theAgent.start()
```

```python
# Q3 and can be Q4.1

import random

class Environment_2x2():
#     ---------
#     | A | B |
#     ---------
#     | D | C |
#     ---------
    def __init__(self):
        # Let's now instantiate Rooms with 'Clean' or 'Dirty' randomly
        self.locationsCondition = {}
        
        # Making sure at most two rooms are Dirty only
        at_most_two_dirt = 0
        # A -> 00   B -> 01   C -> 10   D -> 11
        for room in ['A', 'B', 'C', 'D']:
            
            condition = random.choice(['Clean', 'Dirty'])
            
            if condition == 'Dirty':
                at_most_two_dirt += 1
            if at_most_two_dirt > 2:
                self.locationsCondition[room] = 'Clean'
            
            elif at_most_two_dirt <= 2:
                self.locationsCondition[room] = condition
        

class TableDrivenVacuumAgent_2x2():
#     -----------------
#     |   A   |   B	  |
#     |   🛸> |   	  |
#     -----------------
#     |   D	  |   C   |
#     |		  |		  |
#     -----------------

    def __init__(self, Environment):
        print(Environment.locationsCondition)
        # Instantiate performace measurement
        self.Environment = Environment
        self.Score = 0
        self.vacuumLocation = 'A'   # In start it will be current Location and then changed accordingly
        self.LocationCondition = self.Environment.locationsCondition[self.vacuumLocation]     # In start it will be current Location condition and then changed accordingly
        self.Table = {  # A->B->C-> D-> A
                         ('A','Clean'):'move',
                         ('A','Clean', 'B','Clean'):'turn move',
                         ('A','Clean', 'B','Clean', 'C','Clean'):'turn move',
                         ('A','Clean', 'B','Clean', 'C','Clean', 'D','Clean'):'turn move',
                         ('A','Clean', 'B','Clean', 'C','Clean', 'D','Dirty'):'suck turn move',
                         ('A','Clean', 'B','Clean', 'C','Dirty'):'suck turn move',
                         ('A','Clean', 'B','Clean', 'C','Dirty', 'D','Clean'):'turn move',
                         ('A','Clean', 'B','Clean', 'C','Dirty', 'D','Dirty'):'suck turn move',
                         ('A','Clean', 'B','Dirty'):'suck turn move',
                         ('A','Clean', 'B','Dirty', 'C','Clean'):'turn move',
                         ('A','Clean', 'B','Dirty', 'C','Clean', 'D','Clean'):'turn move',
                         ('A','Clean', 'B','Dirty', 'C','Clean', 'D','Dirty'):'suck turn move',
                         ('A','Clean', 'B','Dirty', 'C','Dirty'):'suck turn move',
                         ('A','Clean', 'B','Dirty', 'C','Dirty', 'D','Clean'):'turn move',

                         ('A','Dirty'):'suck move',
                         ('A','Dirty', 'B','Clean'):'turn move',
                         ('A','Dirty', 'B','Clean', 'C','Clean'):'turn move',
                         ('A','Dirty', 'B','Clean', 'C','Clean', 'D','Clean'):'turn move',
                         ('A','Dirty', 'B','Clean', 'C','Clean', 'D','Dirty'):'suck turn move',
                         ('A','Dirty', 'B','Clean', 'C','Dirty'):'suck turn move',
                         ('A','Dirty', 'B','Clean', 'C','Dirty', 'D','Clean'):'turn move',
                         ('A','Dirty', 'B','Dirty'):'suck turn move',
                         ('A','Dirty', 'B','Dirty', 'C','Clean'):'turn move',
                         ('A','Dirty', 'B','Dirty', 'C','Clean', 'D','Clean'):'turn move',
                          
                        # This was always A->B->C-> D-> A But A->D->C->B->A  A->C->D->B->A  A->D->B->C->A  etc would count upto hundreds of more table driven rules
                                                                                                 }
    
    def _change_location(self, location):   # This function helps in selecting next Location 
        next_locations = {'A':'B', 'B':'C', 'C':'D', 'D':'A'}
        return next_locations[location]
        
    def _map_action_to_actuator(self, action, location):
        if action == 'move':
            print('🛸     Moving to Location {}...'.format( self._change_location(location) ))            
            self.Score -= 1  # -1 for move
        
        elif action == 'turn move':
            print('🔄 🛸  Turning and Moving to Location {}...'.format( self._change_location(location) ))            
            self.Score -= 2   # -1 for turn and -1 for move     
        
        elif action == 'suck move':
            print('🧹     Location {} is Dirty.\n       Location {} has been Cleaned.\n🛸     Moving to Location {}...'.format( location, location, self._change_location(location) ))            
            self.Environment.locationsCondition[location] = 'Clean'
            self.Score += 4   # 5 for suck and -1 for move
            
        elif action == 'suck turn move':
            print('🧹     Location {} is Dirty.\n       Location {} has been Cleaned.\n🔄 🛸  Turning and Moving to Location {}...'.format( location, location, self._change_location(location) ))            
            self.Environment.locationsCondition[location] = 'Clean'
            self.Score += 3   # 5 for suck and -1 for turn and -1 for move
            
        else:  # This would be tirggered when there is a bug in lookup table
            print('.............I am a stupid robot Dumb human.............')
        
        return
            
        
    def start(self):
        print('--Vacuum started at Location A--')
        action = self.Table[('A', self.LocationCondition)]  # As we know we start from upper left corner
        self._map_action_to_actuator(action, 'A')  # Perform action
        
        movedLocation = self._change_location(self.vacuumLocation)  # Now as I have to move to next Location
        movedLocationUnchangedCondition = self.Environment.locationsCondition[movedLocation]
        percept_seq = ['A',self.LocationCondition, movedLocation,movedLocationUnchangedCondition]
        
        self.proceed(movedLocation, movedLocationUnchangedCondition, percept_seq)  # Continue until A is reached again
        
        print('--Came back to A--')  # Reached home hence work finished
        print(self.Environment.locationsCondition)
        print('Performance Measurement: ', self.Score, '\n')
        return self.Score
    
    # proceed handles states like (A,clean, B,clean) etc where Vacuum has moved to next Location
    def proceed(self, Location, LocationUnchangedCondition, percept_seq):
        # As vacuum has already moved from 'A' to next room so now we will alter the variables, could have done in .start() but right now don't want to disturb logic
        self.vacuumLocation = Location
        self.LocationCondition = LocationUnchangedCondition
        
        percept_seq_indexed = tuple(percept_seq)
        action = self.Table[percept_seq_indexed]   
        self._map_action_to_actuator(action, self.vacuumLocation)  # Perform action
        
        movedLocation = self._change_location(self.vacuumLocation)  # Now as I have to move to next Location
        movedLocationUnchangedCondition = self.Environment.locationsCondition[movedLocation]        
        
        if movedLocation != 'A':  # Otherwise it will be going rounds and rounds hence infinity. Case to stop recursion if not satisfied
            percept_seq.append(movedLocation)
            percept_seq.append(movedLocationUnchangedCondition)
            
            self.proceed(movedLocation, movedLocationUnchangedCondition, percept_seq)
        
        return
```
