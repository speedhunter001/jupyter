```python
from graphviz import Digraph

def visualize_tree(tree):
    if tree is None:
        return "Nothing in the tree!"
    def add_nodes_edges(tree, dot=None):
        if dot is None:
            dot = Digraph()
            dot.attr('node', shape='circle')
            dot.node(name=str(tree), label=str(tree.value))
            
        for child in [tree.left, tree.right]:
            if child is not None:
                if child == tree.left:
                    dot.attr('node', shape='circle', style='filled', fillcolor='lightblue')
                if child == tree.right:
                    dot.attr('node', shape='doublecircle', style='filled', fillcolor='seashell')
                dot.node(name=str(child), label=str(child.value))    
                dot.edge(str(tree), str(child))
                dot = add_nodes_edges(child, dot=dot)
        return dot
    
    dot = add_nodes_edges(tree)
    
    display(dot)
```

```python
class Toss:
    def __init__(self, x):
        self.value = x
        self.left = None
        self.right = None
```

```python
def coin_toss(tosses=1, toss=0, root=None):
    
    # First toss so creating '' as root
    if root == None:
        root = Toss('')

    # In this toss assigning H and T as left and right
    root.left = Toss('H')
    root.right = Toss('T')      
        
    # If we have not reached the required number of tosses then do recursive call for another toss
    toss = toss + 1
    if toss != tosses:
        coin_toss(tosses, toss, root.left)
        coin_toss(tosses, toss, root.right)
    
    # Returning whole tree of possibilities
    if root.value == '':
        return root

    
def possibilities(root):
    # If there is no child then return value i.e H or T in a list to iterate over
    if root.left == None and root.right == None:
        return [root.value]
    
    result = []
    if root.left:
        res_left = possibilities(root.left)
        # Iterate over all child possibilities and concatenate value to all child's possibilities
        for poss in res_left:
            result.append(root.value + poss)
    
    if root.right:
        res_right = possibilities(root.right)
        # Iterate over all child possibilities and concatenate value to all child's possibilities
        for poss in res_right:
            result.append(root.value + poss)
    
    # Returning the list of all possible combinations
    return result

def possibilities_dumb(poss=['H', 'T']):
    outcomes = []
    
    for one in poss:
        res = ''
        for two in poss:
            for three in poss:
                for four in poss:
                    for five in poss:
                        for six in poss:
                            for seven in poss:
                                for eight in poss:
                                    for nine in poss:
                                        for ten in poss:
                                            res = one + two + three + four + five + six + seven + eight + nine + ten
                                            outcomes.append(res)
    return outcomes
```

```python
r = coin_toss(10)
```

```python
visualize_tree(r)
```

```python
t1 = %timeit -r10 -n10 -o possibilities(r)
t2 = %timeit -r10 -n10 -o possibilities_dumb() 
```

```python
print("The Awesome recursive solution for stupid coin tosses has time on average  ", t1.average)
print("The Stupid iterative  solution for stupid coin tosses has time on average  ", t2.average)  
# It looks like recursive here is slow :( but still general :)
```
