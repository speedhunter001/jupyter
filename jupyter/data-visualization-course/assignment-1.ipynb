```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("ExcelFormattedGISTEMPData2CSV.csv", index_col='Year')
```

```python
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(data['NHem'], color='red')
ax[1].plot(data['SHem'])
plt.xlabel('Year')
plt.ylabel('Comparison', y=1.20)
plt.savefig('NHem and SHem comparison.png')
plt.show()
```
