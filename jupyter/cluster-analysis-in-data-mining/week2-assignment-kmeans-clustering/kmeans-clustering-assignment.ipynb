```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('./places.txt', names=['longitude', 'latitude'])
df.head()
```

```python
df.plot(kind='scatter', x='longitude', y='latitude')
```

```python
data = df.to_numpy()
print(data)
```

```python
kmeans = KMeans(n_clusters=3).fit(data)
kmeans.labels_  # .labels_ consists of all the cluster labels of the .fit() data
```

```python
df['cluster_no'] = kmeans.labels_
df.head(10)
```

```python
df.plot(kind='scatter', x='longitude', y='latitude', c='cluster_no', cmap='winter')  # without color map one cluster is entirely not visible as cmap is white and black
```

```python
df.loc[:, 'cluster_no'].to_csv('./clusters.txt', sep=' ', header=False)  # as only record no. and cluster no. is required
```
