```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def push(self, value):
        new_node = Node(value)
        
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.next = new_node
        return
    
    def pop(self):
        
        if self.head is None:
            raise Exception("There is nothing in the link list")

        if self.head.next is None:
            value = self.head.value
            self.head = None
            return value
        
        temp = self.head
        while temp.next is not None:
            prev = temp
            temp = temp.next
        value = temp.value
        prev.next = None
        return value
    
    def insert(self, value, index):
        new_node = Node(value)
        
        if self.head is None:
            self.head = new_node
            return
        
        counter = 0
        temp = self.head
        while temp is not None and counter < index:
            prev = temp
            temp = temp.next
            counter = counter + 1
        new_node.next = temp
        prev.next = new_node
        return
    
    def remove(self, value):
        if self.len() == 0:
            raise Exception("The list is empty so nothing to remove")
        
        elif self.head.value == value:
            self.head = self.head.next
            return
         
        try:
            temp = self.head
            while temp.next.value != value:
                temp = temp.next
            temp.next = temp.next.next
            return
        except AttributeError:
            return
        
        
    def len(self):
        counter = 0
        temp = self.head
        while temp is not None:
            temp = temp.next
            counter = counter + 1
        return counter
        
    def get(self, index):
        if self.head is None or index >= self.len():
            raise IndexError("Index out of bound")
            
        counter = 0
        temp = self.head
        while temp is not None and counter < index:
            temp = temp.next
            counter = counter + 1
        return temp.value  
    
    def maximum(self):
        if self.head is None:
            raise IndexError("Link list is empyt")
        
        maximum = self.head.value
        
        temp = self.head
        while temp is not None:
            if maximum < temp.value:
                maximum = temp.value
            temp = temp.next
                
        return maximum
    
    def minimum(self):
        if self.head is None:
            raise IndexError("Link list is empty")
        
        minimum = self.head.value
        
        temp = self.head
        while temp is not None:
            if minimum > temp.value:
                minimum = temp.value
            temp = temp.next
                
        return minimum
    
    def remove_minimum(self):
        if self.head is None:
            raise IndexError("Link List is empty")
        
        minimum = self.minimum()
        
        self.remove(minimum)
        
        return
    
    def third_highest(self):
        if self.head is None:
            raise IndexError("Link List is empty")
        
        first_highest = self.maximum()
        
        def less_than_constraint(constraint):
            temp = self.head
            lesser_than = self.head.value
            while temp is not None:
                if temp.value < constraint and temp.value >= lesser_than:
                    lesser_than = temp.value
                temp = temp.next
            return lesser_than
        
        second_highest = less_than_constraint(first_highest)
        third_highest = less_than_constraint(second_highest)
        
        return third_highest
        
    def most_common(self):
        if self.head is None:
            raise IndexError("Link List is empty")
        
        from collections import Counter
        c = Counter()
        temp = self.head
        
        while temp is not None:
            c[temp.value] = c[temp.value] + 1
            temp = temp.next
        
        return c.most_common()[0][0]
    
    def append_list(self, list_2):
        if self.head is None:
            self.head = list_2.head
            return
        
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.next = list_2.head
        return
    
    def some_operation(self, function):
        if self.head is None:
            raise IndexError("Link List is empty")
        
        temp = self.head
        while temp is not None:
            temp.value = function(temp.value)
            temp = temp.next
        return
    
    def __str__(self):
        l = "["
        temp = self.head
        while temp is not None:
            l = l + str(temp.value) + ", "
            temp = temp.next
        l = l.rstrip(", ")
        l = l + "]"
        return l
```

```python
l1 = LinkedList()
l1.push(4)
l1.push(9)
l1.push(4)
l1.push(36)
print(l1)

print("Maximum is :", l1.maximum())

print("Minimum is :", l1.minimum())

print("Removing minimum")
l1.remove_minimum()
print(l1)

print("Third highest is :", l1.third_highest())

print("Most common is :", l1.most_common())

l2 = LinkedList()
l2.push(0)
l2.push(1)
print(l2)

print("Appending l2 to l1")
l1.append_list(l2)
print(l1)

print("Performing some operation")
from math import sqrt
l1.some_operation(sqrt)
print(l1)
```
