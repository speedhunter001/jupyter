```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def push(self, value):
        new_node = Node(value)
        
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.next = new_node
        return
    
    def pop(self):
        
        if self.head is None:
            raise Exception("There is nothing in the link list")

        if self.head.next is None:
            value = self.head.value
            self.head = None
            return value
        
        temp = self.head
        while temp.next is not None:
            prev = temp
            temp = temp.next
        value = temp.value
        prev.next = None
        return value
    
    def insert(self, value, index):
        new_node = Node(value)
        
        if self.head is None:
            self.head = new_node
            return
        
        counter = 0
        temp = self.head
        while temp is not None and counter < index:
            prev = temp
            temp = temp.next
            counter = counter + 1
        new_node.next = temp
        prev.next = new_node
        return
    
    def remove(self, value):
        if self.len() == 0:
            raise Exception("The list is empty so nothing to remove")
        
        elif self.head.value == value:
            self.head = self.head.next
            return
         
        try:
            temp = self.head
            while temp.next.value != value:
                temp = temp.next
            temp.next = temp.next.next
            return
        except AttributeError:
            return
        
        
    def len(self):
        counter = 0
        temp = self.head
        while temp is not None:
            temp = temp.next
            counter = counter + 1
        return counter
        
    def get(self, index):
        if self.head is None or index >= self.len():
            raise IndexError("Index out of bound")
            
        counter = 0
        temp = self.head
        while temp is not None and counter < index:
            temp = temp.next
            counter = counter + 1
        return temp.value
    
    def reverse_list(self):
        if self.head is None or self.head.next is None:
            return
        
        original_last = self.get_last()
        temp = self.get_last()
        while True:
            if temp is self.head:
                temp.next = None
                self.head = original_last
                #print("head is : ", str(self.head.value))
                break
            
            previous = self._before_this_node(temp)
            #print("previous is : ", str(previous.value))
            temp.next = previous
            temp = temp.next  #next will go to the previous here
        
    def get_last(self):
        if self.head is None:
            raise IndexError("List is empty")
        
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        return temp
        
    def _before_this_node(self, current):
        if self.head is None:
            raise IndexError("List is empty")
        
        temp = self.head
        while temp.next is not current:
            temp = temp.next
        return temp
        
    def __str__(self):
        l = "["
        temp = self.head
        while temp is not None:
            l = l + str(temp.value) + ", "
            temp = temp.next
        l = l.rstrip(", ")
        l = l + "]"
        return l
```
