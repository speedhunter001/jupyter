```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class Singly_Link_List:
    def __init__(self):
        self.head = None
    
    def push(self, value):
        new_node = Node(value)
        
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.next = new_node
        return
    
    def pop(self):
        
        if self.head is None:
            raise Exception("There is nothing in the link list")    
        
        if self.head.next is None:
            value = self.head.value
            self.head = None
            return value
        
        temp = self.head
        while temp.next is not None:
            prev = temp
            temp = temp.next
        value = temp.value
        prev.next = None
        return value
            
    def insert(self, value, index):
        new_node = Node(value)
        
        if self.head is None:
            self.head = new_node
            return
        
        counter = 0
        temp = self.head
        while temp is not None and counter < index:
            prev = temp
            temp = temp.next
            counter = counter + 1
        new_node.next = temp
        prev.next = new_node
        return
        
    def remove(self, index):
        if index >= self.length():
            raise Exception("The element at given index does not exists")
            
        elif self.head is None:
            raise Exception("There is no element in the list")
            
        elif self.head.next is None:
            self.head = None
            return
        
        elif index == 0:
            self.head = self.head.next
            return
            
        counter = 0
        temp = self.head
        while temp is not None and counter < index:
            prev = temp
            temp = temp.next
            counter = counter + 1
        prev.next = temp.next
        return
        
    def length(self):
        counter = 0
        temp = self.head
        while temp is not None:
            temp = temp.next
            counter = counter + 1
        return counter
    
    def get(self, index):
        if self.head is None or index >= self.length():
            raise Exception("The element at given index does not exists")
        
        counter = 0
        temp = self.head
        while temp is not None and counter<index:
            temp = temp.next
            counter = counter + 1
        return temp.value
    
    def __str__(self):
        l = "[ "
        temp = self.head
        while temp is not None:
            l = l + str(temp.value) + ", "
            temp = temp.next
        l = l.rstrip(", ")
        l = l + " ]"
        return l
```

```python
class Doubly_Node:
    def __init__(self, value):
        self.prev = None
        self.value = value
        self.next = None
        
class Double_Link_List:
    def __init__(self):
        self.head = None
    
    def push(self, value):
        new_node = Doubly_Node(value)
        
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp
        return
    
    def pop(self):
        
        if self.head is None:
            raise Exception("There is nothing in the link list")    
        
        if self.head.next is None:
            value = self.head.value
            self.head = None
            return value
        
        temp = self.head
        while temp.next is not None:
            prev = temp
            temp = temp.next
        value = temp.value
        prev.next = None
        return value
    
    def insert(self, value, index):
        new_node = Doubly_Node(value)
        
        if self.head is None:
            self.head = new_node
            return
        
        
        counter = 0
        temp = self.head
        while temp is not None and counter < index:
            prev = temp
            temp = temp.next
            counter = counter + 1
        
        new_node.next = temp
        if temp is not None:
            temp.prev = new_node
        
        prev.next = new_node
        new_node.prev = prev
        return
        
    def remove(self, index):
        if index >= self.length():
            raise Exception("The element at given index does not exists")
            
        elif self.head is None:
            raise Exception("There is no element in the list")
            
        elif self.head.next is None:
            self.head = None
            return
        
        elif index == 0:
            self.head = self.head.next
            self.head.prev = None
            return
            
        counter = 0
        temp = self.head
        while temp is not None and counter < index:
            prev = temp
            temp = temp.next
            counter = counter + 1
        
        prev.next = temp.next
        if temp.next is not None:
            temp.next.prev = prev
        return
    
        
    def get(self, index):
        if self.head is None or index >= self.length():
            raise Exception("The element at given index does not exists")
        
        counter = 0
        temp = self.head
        while temp is not None and counter<index:
            temp = temp.next
            counter = counter + 1
        return temp.value
    
    def length(self):
        counter = 0
        temp = self.head
        while temp is not None:
            temp = temp.next
            counter = counter + 1
        return counter
    
    def __str__(self):
        l = "[ "
        temp = self.head
        while temp is not None:
            l = l + str(temp.value) + ", "
            temp = temp.next
        l = l.rstrip(", ")
        l = l + " ]"
        return l
```
