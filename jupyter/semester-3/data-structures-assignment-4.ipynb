```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class Ring:
    def __init__(self):
        self.head = None

    def push(self, value):
        new_node = Node(value)
        
        if self.head is None:
            self.head = new_node
            self.head.next = self.head
            return
        
        temp = self.head
        while temp.next is not self.head:
            temp = temp.next
        temp.next = new_node
        temp.next.next = self.head
        return
    
    def pop(self):
        
        if self.head is None:
            raise Exception("There is nothing in the link list")

        if self.head.next is self.head:
            value = self.head.value
            self.head = None
            return value
        
        temp = self.head
        while temp.next is not self.head:
            prev = temp
            temp = temp.next
        value = temp.value
        prev.next = self.head
        return value
    
    def get_last(self):
        if self.head is None:
            raise IndexError("List is empty")
        
        temp = self.head
        while temp.next is not self.head:
            temp = temp.next
        return temp
    
    def insert(self, value, index):
        new_node = Node(value)
        
        if self.head is None:
            self.head = new_node
            self.head.next = self.head
            return
        
        if index is 0:
            new_node.next = self.head
            self.head = new_node
            last = self.get_last()
            last.next = self.head
            return
        
        counter = 0
        temp = self.head
        while counter < index:
            prev = temp
            temp = temp.next
            counter = counter + 1
        new_node.next = temp
        prev.next = new_node
        return
    
    def remove(self, value):
        if self.len() == 0:
            raise Exception("The list is empty so nothing to remove")
        
        elif self.head.value == value:
            if self.head.next is self.head:
                self.head = None
                return
            
            else:
                last = self.get_last()
                last.next = self.head.next
                self.head = last.next
                return
         
        temp = self.head
        while temp.next.value != value:
        	if temp.next is self.head and temp.value is not value:
        		return
        	temp = temp.next
        temp.next = temp.next.next
        return

    def remove_at(self, index):
        if self.len() == 0:
            #raise Exception("The list is empty so nothing to remove")
            return

        elif self.head.value == self.get(index):
            if self.head.next is self.head:
                self.head = None
                return
            else:
                last = self.get_last()
                last.next = self.head.next
                self.head = last.next
                return            
        
        temp = self.head
        while temp.next.value != self.get(index):
            temp = temp.next
        temp.next = temp.next.next
        return
    
    def len(self):
        if self.head is None:
            return 0
        
        counter = 0
        temp = self.head
        while temp.next is not self.head:
            temp = temp.next
            counter = counter + 1
        counter = counter + 1
        return counter
        
    def get(self, index):
        if self.head is None:
            raise IndexError("Empty list")
            
        counter = 0
        temp = self.head
        while counter < index:
            temp = temp.next
            counter = counter + 1
        return temp.value
    
    
    def __str__(self):
        if self.head is None:
            return "[]"
        
        l = "["
        temp = self.head
        while temp.next is not self.head:
            l = l + str(temp.value) + ", "
            temp = temp.next
        l = l + str(temp.value) + "]"
        return l


if __name__ == '__main__': 
    r = Ring()
    # r.insert(1, 1)
    r.insert(0, 1)
    r.insert(0, 2)
    r.insert(1, 3)
    r.insert(7, 5)  # different behavrior since it's a ring! 
    print(r)
```
