```python
book_cover = './book-cover.jpg'
next_button = './next.png'
support_sign_out_button = './support-sign-out.png'

ready = pg.confirm("Go to the tab where you have OPENED the book. Make sure that you have scrolled to the top then Press Ok. If you do not want to execute press Cancel")
next_x, next_y = pg.locateCenterOnScreen(next_button)
book_cov = PIL.Image.open(book_cover)
img_no = 0
images = [] 

if ready:
    while True:
        img_path = '{}.png'.format(img_no)
        
        # If cannot scroll 
        if pg.locateCenterOnScreen(support_sign_out_button):
            image = pg.screenshot(imageFilename=None, region=(200,175,1500,800))
            
            # If there is a next page go to that page
            if pg.locateCenterOnScreen(next_button):
                pg.click(next_x, next_y)  
                time.sleep(5)
            # else the book is finished
            else:
                break
        
        # If scroll is possible
        else:    
            image = pg.screenshot(imageFilename=None, region=(200,175,1500,840))
            pg.scroll(-15)
        
        images.append(image)
        img_no += 1

book_cov.save('./book.pdf', save_all=True, append_images=images)
```
