```python
from skimage import color

from skimage.transform import rescale
import PIL
import cv2
import numpy as np

from skimage.filters import threshold_otsu
from skimage.filters import threshold_local
```

```python
image_rescaled = rescale(image, 4, multichannel=True)

fig, ax = plt.subplots(1, 2)
ax[0].imshow(image)
ax[1].imshow(image_rescaled)
```

```python
img = cv2.imread('orig-Image.jpg')
rescaled = cv2.resize(img, None, fx=4, fy=4, interpolation=cv2.INTER_CUBIC)
cv2.imwrite('image-rescaled.png', rescaled)

fig, ax = plt.subplots(1, 2)
ax[0].imshow(img)
ax[1].imshow(rescaled)
```
