#  <font style="color:rgb(34,169,34)">Translation</font>

 Translation is the shifting of objectâ€™s location. If you know the shift in (x,y) direction, let it be (t_x,t_y), you can create the transformation matrix  

###  <font style="color:rgb(34,169,34)">Loading Required Libraries</font>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

###  <font style="color:rgb(34,169,34)">Load color image  </font>

```python
img = cv2.imread('media/M4/f1.jpg',1)
```

###  <font style="color:rgb(34,169,34)">This is how we can get rows,cols of an Image</font>

```python
rows,cols ,channels = img.shape
print(img.shape)
```

###  <font style="color:rgb(34,169,34)">X and Y axis value (in pixels) to shift an object location</font>
A Numpy array of type `np.float32` and pass it into cv2.warpAffine() function

The array with x=1 an y=0 correspond to change in x axis and vice versa the array with x=0 an y=1 correspond to change in y axis

```python
M = np.float32([[1,0,100],[0,1,-200]])
```

###  <font style="color:rgb(34,169,34)">Apply Translation  </font>

> **cv2.[warpAffine](https://docs.opencv.org/4.1.0/da/d54/group__imgproc__transform.html#ga0203d9ee5fcd28d40dbc4a1ea4451983)(src,M,(cols,rows))**

- `src` input Image.
- `M` transformation matrix 
- `cols,rows` Size of the output image, which should be in the form of (width, height). 

`Note`Remember width = number of columns, and height = number of rows.

```python
translated = cv2.warpAffine(img,M,(cols,rows)) # rows and columns are the final window size of the resultant window
```

###  <font style="color:rgb(34,169,34)">Plot Image</font>

```python
plt.figure(figsize=[20,20])
plt.subplot(121);plt.imshow(img[:,:,::-1]);plt.title("Original Image");
plt.subplot(122);plt.imshow(translated[:,:,::-1]);plt.title("translated Image");
```

#  <font style="color:rgb(34,169,34)"> Affine Transformation </font>
 In affine transformation, all parallel lines in the original image will still be parallel in the output image. To find the transformation matrix, we need three points from input image and their corresponding locations in output image. Then `cv2.getAffineTransform` will create a 2x3 matrix which is to be passed to `cv2.warpAffine`.

###  <font style="color:rgb(34,169,34)"> Import Libraries, Load image & get rows,col,channel value </font>

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('media/M4/f4.jpg',1)

rows,cols,ch = img.shape
```

###  <font style="color:rgb(34,169,34)"> Transformation Matrix </font>
We need 2 arrays consist of three points. 
1. 1st from input image
2. 2nd for their corresponding locations in output image.

```python
pts1 = np.float32([[0,0],[cols,0],[0,rows]])              # collect 3 points
pts2 = np.float32([[10,100],[200,50],[50,350]])     # modify them
```

Calculates an affine transform from three pairs of the corresponding points.

The function calculates the 2Ã—3 matrix of an affine transform 
> **cv.[getAffineTransform](https://docs.opencv.org/4.1.0/da/d54/group__imgproc__transform.html#ga47069038267385913c61334e3d6af2e0)(src, dst)**

**Parameters**
- `src`	Coordinates of triangle vertices in the source image.
- `dst`	Coordinates of the corresponding triangle vertices in the destination image.

```python
M = cv2.getAffineTransform(pts1,pts2)
```

###  <font style="color:rgb(34,169,34)"> Apply Affine Transformation  </font>

```python
dst = cv2.warpAffine(img,M,(cols,rows))
```

###  <font style="color:rgb(34,169,34)"> Show image </font>

```python
plt.figure(figsize=[20,20])
plt.subplot(121);plt.imshow(img[:,:,::-1]);plt.title("Original Image");
plt.subplot(122);plt.imshow(dst[:,:,::-1]);plt.title("Affined Image");
```

#  <font style="color:rgb(34,169,34)">Perspective Transformation </font>
Straight lines will remain straight even after the transformation. To find this transformation matrix, you need 4 points on the input image and corresponding points on the output image. Among these 4 points, 3 of them should not be collinear. Then transformation matrix can be found by the function cv.getPerspectiveTransform. Then apply cv.warpPerspective with this 3x3 transformation matrix.

###  <font style="color:rgb(34,169,34)"> Import Libraries, Load image & get rows,col,channel value </font>

```python
import cv2
import numpy as np
img = cv2.imread('media/M4/sidebook1.jpg')

rows,cols,ch = img.shape
```

###  <font style="color:rgb(34,169,34)"> Transformation Matrix </font>
We need 2 arrays consist of four points. 
1. 1st from input image
2. 2nd for their corresponding locations on output image.

`note`  Among these 4 points, 3 of them should not be collinear

```python
pts1 = np.float32([[125,127],[246,66],[235,270],[357,210]])
pts2 = np.float32([[0,0],[cols,0],[0,rows],[cols,rows]])
```

###  <font style="color:rgb(34,169,34)"> Calculates a perspective transform</font>

> **cv.[getPerspectiveTransform](https://docs.opencv.org/4.1.0/da/d54/group__imgproc__transform.html#ga20f62aa3235d869c9956436c870893ae)(src, dst)**

**Parameters**
- `src`	Coordinates of quadrangle vertices in the source image.
- `dst`	Coordinates of the corresponding quadrangle vertices in the destination image.


```python
M = cv2.getPerspectiveTransform(pts1,pts2)
```

###  <font style="color:rgb(34,169,34)"> Apply Perspective Transformation & Plot Image</font>

```python
dst = cv2.warpPerspective(img,M,(cols,rows))
%% matplotlib inline
plt.figure(figsize=[20,20])
plt.subplot(121);plt.imshow(img[:,:,::-1]);plt.title("Original Image");
plt.subplot(122);plt.imshow(dst[:,:,::-1]);plt.title("Transformed Image");
```

#  <font style="color:rgb(34,169,34)"> Erosion </font>
 The basic idea of erosion is just like soil erosion only, it erodes away the boundaries of foreground object (Always try to keep foreground in white). So what it does? The kernel slides through the image (as in 2D convolution). A pixel in the original image (either 1 or 0) will be considered 1 only if all the pixels under the kernel is 1, otherwise it is eroded (made to zero). <br><br>So what happends is that, all the pixels near boundary will be discarded depending upon the size of kernel. So the thickness or size of the foreground object decreases or simply white region decreases in the image. It is useful for removing small white noises

###  <font style="color:rgb(34,169,34)"> Import Libraries & Load Image </font>

```python
import cv2
import numpy as np

img = cv2.imread('media/M4/bwtext.png',0)
```

###  <font style="color:rgb(34,169,34)"> Create a Kernel </font>
I would use a 7x7 kernel with full of ones.

```python
kernel = np.ones((7,7),np.uint8)                           #kernel window size 7x7
```

###  <font style="color:rgb(34,169,34)"> Apply Erosion </font>

> **cv.[erode](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#gaeb1e0c1033e3f6b891a25d0511362aeb)(src, kernel)**

**Parameters**
- `src`	input image; the number of channels can be arbitrary, 
- `kernel` structuring element used for erosion

Third input is optional i.e the number of iteration we want to perform on our image.

```python
erosion = cv2.erode(img,kernel,iterations = 2)
```

###  <font style="color:rgb(34,169,34)"> Show Image </font>

```python
plt.figure(figsize=[20,20])
plt.subplot(121);plt.imshow(img[:,:],cmap='gray');plt.title("Original Image");
plt.subplot(122);plt.imshow(erosion[:,:],cmap='gray');plt.title("Eroded Image");
```

#  <font style="color:rgb(34,169,34)"> Dilation </font>
It is just opposite of erosion. Here, a pixel element is â€˜1â€™ if atleast one pixel under the kernel is â€˜1â€™. So it increases the white region in the image or size of foreground object increases. Normally, in cases like noise removal, erosion is followed by dilation. Because, erosion removes white noises, but it also shrinks our object. So we dilate it. Since noise is gone, they wonâ€™t come back, but our object area increases. It is also useful in joining broken parts of an object.

> **cv.[dilate](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#ga4ff0f3318642c4f469d0e11f242f3b6c)(src, kernel)**

**Parameters**
- `src`	input image; the number of channels can be arbitrary, 
- `kernel` structuring element used for erosion

Third input is optional i.e the number of iteration we want to perform on our image.

```python
import cv2
import numpy as np

img = cv2.imread('media/M4/bwtext.png',0)

kernel = np.ones((7,7),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 2)

plt.figure(figsize=[20,20])
plt.subplot(121);plt.imshow(img[:,:],cmap='gray');plt.title("Original Image");
plt.subplot(122);plt.imshow(dilation[:,:],cmap='gray');plt.title("dialated Image");
```

#  <font style="color:rgb(34,169,34)"> Opening </font>
Opening is just another name of erosion followed by dilation. It is useful in removing noise, as we explained above. Here we use the function:
> **cv2.[morphologyEx](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#ga67493776e3ad1a3df63883829375201f)(src, op, kernel)**

**Parameters**
- `src`	Source image. The number of channels can be arbitrary. 
- `op`	Type of a morphological operation, see MorphTypes
- `kernel`	Structuring element. It can be created using getStructuringElement.

```python
import cv2
import numpy as np
img = cv2.imread('media/M4/opening1.png',0)

img2= img.copy()
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img2,kernel,iterations = 1)
dilation = cv2.dilate(erosion,kernel,iterations = 1)

opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

plt.figure(figsize=[10,10])
plt.subplot(131);plt.imshow(img[:,:],cmap='gray');plt.title("Original Image");
plt.subplot(132);plt.imshow(dilation[:,:],cmap='gray');plt.title("Custom Opening Image");
plt.subplot(133);plt.imshow(opening[:,:],cmap='gray');plt.title("Opening Image");
```

#  <font style="color:rgb(34,169,34)"> Closing </font>
Closing is reverse of Opening, Dilation followed by Erosion. It is useful in closing small holes inside the foreground objects, or small black points on the object.

```python
img = cv2.imread('media/M4/closing1.png',0)
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)
erosion = cv2.erode(dilation,kernel,iterations = 1)

closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=[12,10])
plt.subplot(131);plt.imshow(img[:,:],cmap='gray');plt.title("Original Image");
plt.subplot(132);plt.imshow(erosion[:,:],cmap='gray');plt.title("Custom Closing Image");
plt.subplot(133);plt.imshow(closing[:,:],cmap='gray');plt.title("Opening Image");
```

#  <font style="color:rgb(34,169,34)"> Morphological Gradient </font>
It is the difference between dilation and erosion of an image.<br>The result will look like the outline of the object.

```python
img = cv2.imread('media/M4/bwtext.png',0)
kernel = np.ones((2,2),np.uint8)  #increase the kernel size

gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

plt.figure(figsize=[30,20])
plt.subplot(121);plt.imshow(img[:,:],cmap='gray');plt.title("Original Image");
plt.subplot(122);plt.imshow(gradient[:,:],cmap='gray');plt.title("Gradient Image");
```

## <span style="background-color: #FFFF00">Sobel & Laplacian Derivatives Or Crude Edge Detection.</span>



Sobel operators is a joint Gausssian smoothing plus differentiation operation, so it is more resistant to noise. You can specify the direction of derivatives to be taken, vertical or horizontal
 Laplacian is the sum of x and y sobel partial derivatives taken at kernel =1 and then applied a certain 2d filter.
 
 > **cv2.[Laplacian](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#gad78703e4c8fe703d479c1860d76429e6)(src, ddepth)**
 
**Parameters**
- `src:`	Source image.
- `ddepth:`	Desired depth of the destination image.

> **cv2.[Sobel](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#gacea54f142e81b6758cb6f375ce782c8d)(src, ddepth, dx, dy, ksize)**

**Parameters**

- `src`	input image.
- `ddepth`	output image depth, see combinations; in the case of 8-bit input images it will result in truncated derivatives.
- `dx`	order of the derivative x.
- `dy`	order of the derivative y.
- `ksize` size of the extended Sobel kernel; it must be 1, 3, 5, or 7.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img2 = cv2.imread('media/M4/whiteblock.jpg',0)


img = img2
laplacian = cv2.Laplacian(img,cv2.CV_64F)                 #Laplacian is sum of x and y axis
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=-1)       #1,0 means on x axis
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)       #0,1 means on y axis
print(sobelx.dtype)

plt.figure(figsize=[10,10])
plt.subplot(221);plt.imshow(img2[:,:],cmap='gray');plt.title("Original Image");
plt.subplot(222);plt.imshow(laplacian[:,:],cmap='gray');plt.title("laplacian Image");
plt.subplot(223);plt.imshow(sobelx[:,:],cmap='gray');plt.title("sobelx Image");
plt.subplot(224);plt.imshow(sobely[:,:],cmap='gray');plt.title("sobely Image");
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

try:
    cap.release()
except:
    pass


cap = cv2.VideoCapture(0)
while 1:
    ret,img = cap.read()
    laplacian = cv2.Laplacian(img,cv2.CV_64F,ksize=13)
    sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=13)
    sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=13)
    k = cv2.waitKey(1)
    if k & 0xFF == 27:
        break
    cv2.imshow('laplacian',laplacian)
    cv2.imshow('sobelx',sobelx)
    cv2.imshow('sobely',sobely)


cap.release()
cv2.destroyAllWindows()    
```

## <mark>Canny edge Detection (A classic edge detection alogrithim)</mark>
he function finds edges in the input image and marks them in the output map edges using the Canny algorithm. The smallest value between threshold1 and threshold2 is used for edge linking. The largest value is used to find initial segments of strong edges

> **cv2.[Canny](https://docs.opencv.org/4.1.0/dd/d1a/group__imgproc__feature.html#ga2a671611e104c093843d7b7fc46d24af)(img, threshold1, threshold2)**

- `img`: source image
- `threshold1:`	first threshold for the hysteresis procedure.
- `threshold2:`	second threshold for the hysteresis procedure.

 **Note** Tune the min and max values in cv2.canny() function to get the results you want

```python
try:
    cap.release()
    cv2.destroyAllWindows()      

except:
    pass
cap = cv2.VideoCapture(0)
cv2.namedWindow('Canny',cv2.WINDOW_NORMAL)

while 1:
    ret,img = cap.read()
    cv2.flip(img,1)
    img = cv2.flip( img, 1 )
    edges = cv2.Canny(img,100,200)


    k = cv2.waitKey(1)
    if k & 0xFF == 27:
        break
    cv2.imshow('Canny',edges)
    cv2.imshow('orignal',img)


cap.release()
cv2.destroyAllWindows() 
```

## Image Pyramids Changing Image Resolution & Laplacian pyramid

> **cv2.[pyrDown](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#gaf9bba239dfca11654cb7f50f889fc2ff)(img)**

Blurs an image and downsamples it

> **cv2.[pyrUp](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#gada75b59bdaaca411ed6fee10085eb784)(img)**

Upsamples an image and then blurs it.

**Parameters**
- `img:`	input image.

```python
img = cv2.imread('media/M4/Trump1.jpg')
img = cv2.resize(img, (500, 500))
print('oringinal image shape' , img.shape)

lower_reso = cv2.pyrDown(img)
lower_reso2 = cv2.pyrDown(lower_reso)

higher_reso2 = cv2.pyrUp(lower_reso2)

print('lower reso shape',lower_reso.shape)
print('lower reso 2 shape',lower_reso2.shape)

higher_reso = cv2.pyrUp(lower_reso)

print('see that lower reso and higher reso2 shapes are same',lower_reso.shape,higher_reso2.shape)
Lub = cv2.subtract(lower_reso,higher_reso2 ) # this is laplacian pyramid


print('see that higher reso and img shapes are same', higher_reso.shape,img.shape)


plt.figure(figsize=[15,10])
plt.subplot(231);plt.imshow(img[:,:,::-1]);plt.title("Original Image");
plt.subplot(232);plt.imshow(lower_reso[:,:,::-1]);plt.title("one level down Image");
plt.subplot(233);plt.imshow(lower_reso2[:,:,::-1]);plt.title("Two levels down Image");  
plt.subplot(234);plt.imshow(Lub[:,:,::-1]);plt.title("laplacian Image");  
plt.subplot(235);plt.imshow(higher_reso[:,:,::-1]);plt.title("level up of one level down Image");
plt.subplot(236);plt.imshow(higher_reso2[:,:,::-1]);plt.title("level up of two level down Image");
```

```python
#laplacian pyramids are not the same as laplacian derivatives
A = cv2.imread('media/M4/Trump1.jpg')
laplacian = cv2.Laplacian(A,cv2.CV_64F)
edges = cv2.Canny(A,100,300)
   
plt.figure(figsize=[20,15])
plt.subplot(131);plt.imshow(edges[:,:],cmap='gray');plt.title("plain canny Image");
plt.subplot(132);plt.imshow(laplacian[:,:],cmap='gray');plt.title("Plain laplacian Image");
plt.subplot(133);plt.imshow(Lub[:,:,::-1],cmap='gray');plt.title("laplacian derivative Image");
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.resize(cv2.imread('media/M4/apple.jpg',1), (320, 320))
img2 =cv2.resize(cv2.imread('media/M4/orange.jpg',1), (320, 320))

rows,cols,chan = img.shape

M = np.float32([[1,0,0],[0,1,-17]])
dst = cv2.warpAffine(img,M,(cols,rows),borderMode=1)
merged= cv2.addWeighted(dst,0.5,img2,0.5,0)

plt.figure(figsize=[15,10])
plt.subplot(131);plt.imshow(dst[:,:,::-1]);plt.title("Image 1");
plt.subplot(132);plt.imshow(img2[:,:,::-1]);plt.title("Image 2");
plt.subplot(133);plt.imshow(merged[:,:,::-1]);plt.title("merged Image");  

cv2.imwrite('media/M4/applem.jpg',dst)
```

## Pyramid Image Merging Application (step by step)

```python
import cv2
import numpy as np,sys

A = cv2.imread('media/M4/applem.jpg')
B = cv2.imread('media/M4/orange.jpg')
A = cv2.resize(A, (320, 320))
B = cv2.resize(B, (320, 320))
print(A.shape,B.shape)

# generate Gaussian pyramid for  
# we go 6 layers down
G = A.copy()
gpA = [G]
for i in range(6):
    G = cv2.pyrDown(G)
    gpA.append(G)
for x in range(len(gpA)):
    cv2.imshow(str(x),gpA[x])


cv2.waitKey(0)
cv2.destroyAllWindows()          
```

## Pyramid Image Merging Application (Full Code)

```python
import time
import cv2
import numpy as np,sys
A = cv2.imread('media/M4/imrankhanface1.jpg')

B = cv2.imread('media/M4/tomface1.jpg')

#A = cv2.imread('apple.png')
#B = cv2.imread('orange.png')
A = cv2.resize(A, (640, 640))
B = cv2.resize(B, (640, 640))
print(A.shape,B.shape)

# generate Gaussian pyramid for A
G = A.copy()
gpA = [G]
for i in range(6):
    G = cv2.pyrDown(G)
    gpA.append(G)

# generate Gaussian pyramid for B
G = B.copy()
gpB = [G]
for i in range(6):
    G = cv2.pyrDown(G)
    gpB.append(G)

    
  
#generate Laplacian Pyramid for A
lpA = [gpA[5]]
for i in range(5,0,-1):
    GE = cv2.pyrUp(gpA[i])
    print(gpA[i-1].shape,GE.shape)
    L = cv2.subtract(gpA[i-1],GE)
    lpA.append(L)

# generate Laplacian Pyramid for B
lpB = [gpB[5]]
for i in range(5,0,-1):
    GE = cv2.pyrUp(gpB[i])
    L = cv2.subtract(gpB[i-1],GE)
    lpB.append(L)  

# Now add left and right halves of images in each level
LS = []
for la,lb in zip(lpA,lpB):
    rows,cols,dpt = la.shape
   # ls = cv2.add(la,lb)
    #ls = cv2.addWeighted(la,0.5,lb,0.5,0)

    ls = np.hstack((la[:,0:int(cols/2)], lb[:,int(cols/2):]))
    LS.append(ls)

# now reconstruct
ls_ = LS[0]
for i in range(1,6):
    ls_ = cv2.pyrUp(ls_)
    ls_ = cv2.add(ls_, LS[i])

# image with direct connecting each half
real = np.hstack((A[:,:int(cols/2)],B[:,int(cols/2):]))

plt.figure(figsize=[20,10])
plt.subplot(131);plt.imshow(real[:,:,::-1]);plt.title("Direct_blending");
plt.subplot(132);plt.imshow(ls_[:,:,::-1]);plt.title("Pyramid_blending");
```

```python

def warper(A,B):    
    A = cv2.resize(A, (384, 384))
    B = cv2.resize(B, (384, 384))
    
    G = A.copy()
    gpA = [G]
    for i in range(6):
        G = cv2.pyrDown(G)
        gpA.append(G)

    G = B.copy()
    gpB = [G]
    for i in range(6):
        G = cv2.pyrDown(G)
        gpB.append(G)
        
    lpA = [gpA[5]]
    for i in range(5,0,-1):
        GE = cv2.pyrUp(gpA[i])
        L = cv2.subtract(gpA[i-1],GE)
        lpA.append(L)

    lpB = [gpB[5]]
    for i in range(5,0,-1):
        GE = cv2.pyrUp(gpB[i])
        L = cv2.subtract(gpB[i-1],GE)
        lpB.append(L)  

    LS = []
    for la,lb in zip(lpA,lpB):
        rows,cols,dpt = la.shape
        #ls = cv2.addWeighted(la,0.1,lb,0.7,0)
        ls = np.hstack((la[:,0:int(cols/2)], lb[:,int(cols/2):]))
        LS.append(ls)

    ls_ = LS[0]
    for i in range(1,6):
        ls_ = cv2.pyrUp(ls_)
        ls_ = cv2.add(ls_, LS[i])
    return ls_    
```

## Pyramid Image Merging Application 2 Live Merging of Faces

```python
face_cascade = cv2.CascadeClassifier('media/M4/haarcascade_frontalface_default.xml')

try:
    cap.release()
except:
    pass

#gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cap = cv2.VideoCapture(1)
cv2.namedWindow('imgm', cv2.WINDOW_NORMAL)
cv2.namedWindow('img', cv2.WINDOW_NORMAL)

while 1:
    
        ret, img = cap.read()
        img = cv2.resize(img, (480, 480)) 

        
        if ret:
            img = cv2.flip( img, 1 )
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            

            #print(gray.shape)
            faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        
            for (x,y,w,h) in faces:
                    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
                    font = cv2.FONT_HERSHEY_SIMPLEX
                    if x+ int(w/2) > 240:
                         face2 = img[y:y+h,x:x+w]
                            
                         cv2.putText(img,'2nd Face Detected',(x,y+h+15), font, 0.5, (0,0,255), 2, cv2.LINE_AA)   

                    else:
                        face1 = img[y:y+h,x:x+w]

                        cv2.putText(img,'1st Face Detected',(x,y+h+15), font, 0.5, (0,0,255), 2, cv2.LINE_AA) 
                    try:
                      combined = warper(face1,face2)   
                      cv2.imshow('imgm',combined)
                    except:
                        pass

        
        cv2.imshow('img',img)

        k = cv2.waitKey(1) 
        if k == ord('a'):
            break
   
cap.release()
cv2.destroyAllWindows()
```
