# Image Segmentation 

Image Segmentation can be called a fancy object detection. Image Segmentation what you do after you have compeletely understood the concepts of object. But what if you were required to construct pixel-wise masks for every object in an image? Or what if in simple words you were said to segment each image seperately? Well in that case the image segmentation comes in. 
Why can't we just take the image and put rectangles on the image and call it image segmentation?

## Instance Segmentation
Instance segmentation, the automatic delineation of different objects appearing in an
image.Instance segmentation can be formulated as the conjunction between semantic segmentation and object detection.

We can see in the example given below why object localization and object detection fails compared to **Instance Segmentation**. You can see that instance segmentation is a way better option to go with when we need our ROI(region of interest) in a picture. Also instance segmentation gives us the clarity of how dense an image is by outlining the end of one object/person/thing and start of next object/person/thing.

*-Instance segmentation algorithms, compute a pixel-wise mask for every object in the image, even if the objects are of the same class label. Here in the given example you can see that each of the person has their own unique color, implying that our instance segmentation algorithm not only localized each individual but predicted their boundaries as well-.*

## Process of Instance Segmentation

Now that you are a little familiar with all the pre-trained CNN, now we can talk a little more in-depth about how and what is happening in the background. The process of our code is going to be the same as we have been doing, we shall load our NN model weights and configuration, pre-process our data, re-configure the output using OpenCV and lastly get amazing instance segmentations.

But what's going on behind the scenes here? Well, Instance Segmentation is formally known as RCNN Masking, RCNN stands for Region with CNN feature, meaning that RCNN is used to classify our images and segment them with some masks. RCNN doesn't use the entire picture for the input, so for instance, the example we gave above had people in it and maybe their were some other things in it, but maybe we just wanted to localize the people and them segment them, then why must we process and classify the entire images? Why don't we just take a part of that image and classify the ROI we need in it! and that is the entire use of RCNN's here, rather than processing entire images and extracting features, it just targets the ROI in the pictures and takes out feature from them to learn/train from.

````python
import numpy as np
import random
import time
import cv2
import os
````

### Loading the Labels

Here we load the labels, load some colors to work with, and we also load my ```np.random.seed(50)``` we do that, so when you code we both would have the same random values, even though they are random, we both would get the same random values.

````python
# load the COCO class labels our Mask R-CNN was trained on

LABELS = open("object_detection_classes_coco.txt").read().strip().split('\n')
 
    
# load the set of colors that will be used when visualizing a given instance segmentation
# We use the same seed so that both our codes match even for a rand func
np.random.seed(50)

# Randomly select the colors that shall segment our image
COLORS = np.random.randint(0, 255, size=(10, 3))

# Analyzing our defined data
print("First Three Labels:",LABELS[:3])
print("First Three Colors: \n",COLORS[:3])
print("Length of all Labels:",len(LABELS))
````

### Loading the RCNN

Here we load the model weights and configuration. We have been doing and using the same cv2 function called ```cv2.dnn.readNetFromCaffee()``` but here we use ```cv2.dnn.readNetFromTensorflow()``` since the model must have been trained in tensorflow, it doesn't matter where the model was trained, OpenCV support about all the CNN models.

````python
# derive the paths to the Mask R-CNN weights and model configuration
weightsPath = 'frozen_inference_graph.pb'
configPath = 'mask_rcnn_inception_v2_coco_2018_01_28.pbtxt'

# load our Mask R-CNN trained on the COCO dataset (90 classes)
# from disk
print("[INFO] loading Mask R-CNN from disk...")
net = cv2.dnn.readNetFromTensorflow(weightsPath, configPath)
net.setPreferableTarget( cv2.dnn.DNN_TARGET_OPENCL)
print("[DONE] R-CNN loaded successfully")
````

### Testing RCNN 

Before taking our feed forward RCNN to the next level and start using it mask pictures, we first try whether everything is loaded perfectly or not. Which is why we test this entire code here. 

We see that in the ```(boxes, masks)``` variable the RCNN returns some values these are the predicted places where our images segmentation is happening. How did we know the ```"detection_out_final", "detection_masks"``` names from the list, we these are initialized by the creators of this NN and that these were seen from their documentation.

````python
# load our input image and grab its spatial dimensions
image = cv2.imread('man.jpg')
#image = cv2.resize(image, (1000, 500))
# Height and Width variables 
(H, W) = image.shape[:2]

# Now, we initialize a blob from our input image
blob = cv2.dnn.blobFromImage(image, swapRB=True, crop=False)

# We set the inputs, initialize the timestamps and finally make a forward pass in our RCNN
# The RCNN returns 2 values and they're stored in boxes and masks
net.setInput(blob)
start = time.time()
(boxes, masks) = net.forward(["detection_out_final", "detection_masks"])
end = time.time()
 
    

# Paste the outputs
print("[INFO] Mask R-CNN took {:.6f} seconds".format(end - start))
print("[INFO] boxes shape: {}".format(boxes.shape))
print("[INFO] masks shape: {}".format(masks.shape))
````

### Drawing the Boxes and Masks 

Now that we have made sure that our RCNN is working just fine, we shall move forward to drawing those boxes and masks that were predicted by our model. For doing so we shall use openCV for drawing our ROI. 

This code snippet will find the ROI in this image once at a time and then you will be able to see them once window at a time. 

````python
#this will act as a flag
vis = 1

# here we iterate over all the objects detected in our image
for i in range(0, boxes.shape[2]):
    
    # Here we will take out the classID's of all the predicted objects and 
    # their confidence (i.e., probability) 
    classID = int(boxes[0, 0, i, 1])
    confidence = boxes[0, 0, i, 2]
    
    # Here we put an IF-ELSE condition to make sure all the correct
    # predictions are masked 
    if confidence > 0.5:
        
        
        # clone our original image so we can draw on it
        # we do this so that our original image is untouched
        clone = image.copy()
        
        '''Here we, scale the bounding box coordinates back relative to the size of the image 
        and then compute the width and the heightof the bounding box'''
        box = boxes[0, 0, i, 3:7] * np.array([W, H, W, H])
        (startX, startY, endX, endY) = box.astype("int")
        boxW = endX - startX
        boxH = endY - startY
        
        
        '''image segmentation requires that we find all pixels where an object is present. Thus, we’re going to place a transparent 
        overlay on top of the object to see how well our algorithm is performing. In order to do so, we’ll calculate a mask:'''
        # Here we, take out the pixel-wise segmentation for the object
        mask1 = masks[i, classID]
        
        # Resize the mask such that it's the same dimensions of the bounding box
        # resizing the 15 by 15 mask to the roi size
        mask2 = cv2.resize(mask1, (boxW, boxH),interpolation=cv2.INTER_NEAREST) 
        
        # and then finally threshold to create a *binary* mask
        # if you set this value to 0 then it will give you the mask of the whole box , the threshold filters
        mask = (mask2 > 0.3)
        
        '''Here we take out the highest probabilty pixels, which we are sure are ROI
            now the mask that you have is a binary mask extract the ROI of the image'''
        roi = clone[startY:endY, startX:endX]
    
    
        if vis > 0:
            # here we, set greater than 0 used for visualizing masks
            # then, convert the mask from a boolean to an integer mask with
            # to values: 0 or 255, then apply the mask
            visMask = (mask * 255).astype("uint8") # turning the binary mask to a black and white mask
            instance = cv2.bitwise_and(roi, roi, mask=visMask)

            # output the extracted ROI, the mask, along with the
            # segmented instance
            cv2.imshow("ROI", roi)
            cv2.imshow("Mask", visMask)
            cv2.imshow("Segmented", instance)
            
            
            # now, extract 'only' the masked region of the ROI by passing
            # in the boolean mask array as our slice condition
            roi = roi[mask]

            # randomly select a color that will be used to visualize this
            # particular instance segmentation then create a transparent
            # overlay by blending the randomly selected color with the ROI
            color = random.choice(COLORS)
            blended = ((0.5 * color) + (0.4 * roi)).astype("uint8")

            # store the blended ROI in the original image
            clone[startY:endY, startX:endX][mask] = blended
            
            # draw the bounding box of the instance on the image
            color = [int(c) for c in color]
            cv2.rectangle(clone, (startX, startY), (endX, endY), color, 2)

            # draw the predicted label and associated probability of the
            # instance segmentation on the image
            text = "{}: {:.3f}".format(LABELS[classID], confidence)
            cv2.putText(clone, text, (startX, startY - 5),cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

            # show the output image
            cv2.imshow("Output", clone)
            cv2.waitKey(0)
            
cv2.destroyAllWindows()            
````

### Drawing Boxes and Masks(Video)

What we have performed in the above code, we will do this for each frame of the video that we have initilized, here we will be giving the video on our own. 

#### Imports 

````python
import numpy as np
import random
import time
import cv2
import os
````

````python

frameskip = 6

# The have a video in our directory named example.mp4
filename='example.mp4'

# Here we take all the labels 
LABELS = open('object_detection_classes_coco.txt').read().strip().split("\n")

# we have the same seed as that I have, so that we both could
# have the same random results
np.random.seed(50)

# We load some colors to mask our predicted ROI's
COLORS = np.random.randint(0, 255, size=(100, 3))

# Initializing weights and model configurations
weightsPath = 'frozen_inference_graph.pb'
configPath = 'mask_rcnn_inception_v2_coco_2018_01_28.pbtxt'
net = cv2.dnn.readNetFromTensorflow(weightsPath, configPath)

# We initialize our video
cap = cv2.VideoCapture(filename)

#Total number of frames in our Video
prop = cv2.CAP_PROP_FRAME_COUNT
firstframe = False

# Format of Video
fourcc = cv2.VideoWriter_fourcc(*'XVID')

# Here, we initialize how we want our output, see the OpenCV 
# docs for further formats of output
out = cv2.VideoWriter('maskout.avi',fourcc, 5.0, (640,480)) #fps, width, height

# Since our process of segmentation is going to take 
# quite some time depending on your PC's GPU, hence we 
# have estimated the time it will take to segment some 
# portions of your video
initial = 0
total = int(cap.get(prop))
print("[INFO] {} total frames in video".format(total))

'''This while Loop keeps going infinitely until
the if-condition doesnt break the loop'''
print('started')
while 1:
    ret, image = cap.read()  
    if not ret:
        break
    initial += 1
    if initial is not frameskip:
        continue
    
    initial = 0  
    
    # Taking out the prediction height and width of Segmented image for masking 
    (H, W) = image.shape[:2]
    
    # Pre-Processing the image
    blob = cv2.dnn.blobFromImage(image, swapRB=True, crop=False)
    
    # Setting the input
    net.setInput(blob)
    
    # Marking timestamps and then making a forward pass in RCNN
    start = time.time()
    (boxes, masks) = net.forward(["detection_out_final", "detection_masks"])
    end = time.time()
    
    # Looping over each of the predicted classID and confidence(i.e probablity)
    # using OpenCV to mask our image
    for i in range(0, boxes.shape[2]):
        classID = int(boxes[0, 0, i, 1])  # Taking our classID
        confidence = boxes[0, 0, i, 2]    # Taking probablity 
        if confidence > 0.5:
            clone = image.copy()
            
            '''Masking our predicted from here'''
            # Here we, scale the bounding box coordinates back relative to the size of the image
            # and then compute the width and the heightof the bounding box
            box = boxes[0, 0, i, 3:7] * np.array([W, H, W, H])
            (startX, startY, endX, endY) = box.astype("int")
            boxW = endX - startX
            boxH = endY - startY
            
            # image segmentation requires that we find all pixels where an object is present. 
            # Thus, we’re going to place a transparent. overlay on top of the object to see how well our algorithm is performing.
            # In order to do so, we’ll calculate a mask:
            mask1 = masks[i, classID]
            mask2 = cv2.resize(mask1, (boxW, boxH),interpolation=cv2.INTER_NEAREST)
            mask = (mask2 > 0.3)
            
            # Here we take out the highest probabilty pixels, which we are sure are ROI
            # now the mask that you have is a binary mask extract the ROI of the image
            roi = clone[startY:endY, startX:endX]      
            roi = roi[mask]
            color = COLORS[classID]
            color = [int(c) for c in color]
            
            # Here we’ve blended  our roi  with color and store  it in the original blended(variable) 
            # effectively creating a colored transparent overlay
            blended = ((0.4* np.asarray(color)) + (0.6*roi)).astype("uint8")
            
            # Saving the Blended ROI in the Cloned Image Frame
            clone[startY:endY, startX:endX][mask] = blended
            
            # Coloring the bounding box
            color = [int(c) for c in color]
            cv2.rectangle(clone, (startX, startY), (endX, endY), color, 2)
            
            # Drawing the BOX on predicted object/thing
            text = "{}: {:.3f}".format(LABELS[classID], confidence)
            cv2.putText(clone, text, (startX, startY - 5),cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

# We calcualte the time it shall take to complete this segmentaion of
# Video, we do this because it is GPU/CPU draining process depending on 
# your PC it will vary each time!
    if firstframe ==False:
            total = total / frameskip
            elap = (end - start)
            print("single frame took {:.2f} seconds".format(elap))
            print("estimated time to finish: {:.2f} minutes".format((elap * total)/60))
            firstframe = True
    
    out.write(clone)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

out.release()            
cv2.destroyAllWindows()            
cap.release()
    
print('all done')
````
