## <font style="color:rgb(134,19,348)">  What is Image Segmentation?  </font> 

Image Segmentation can also be called as fancy object detection. if you undertstood the concept of obejct detection clearly than image segmentation will be so much easier for you.Lets say if you are required to construct pixel-wise masks for every object detection in an image? Or what if in simple words you were said to segment each image object seperately? that's sounds like hectic but in our case we are relax because we know the technique of image segmentation. 

**So as we know the image contain lots of pixels, but for image segmentation we are just interested into grouping those pixels which have similar attributes.**


## <font style="color:rgb(134,19,348)">   Types of Image Segmentation</font> 

Although there are lots of technique for Image Segmentation but we divide into narrow circle.where most of categories lies.

**Two types of Image segmentation:**

- **`Semantic segmentation:`**	 
- **`Instance segmentation:`**	 



### <font style="color:rgb(134,19,348)">   Semantic Segmentation</font> 

In Semantic Segmentation every pixel of object belong to a particular class are represented by the same color. It focus on classifying all the people as a single instance 

### <font style="color:rgb(134,19,348)">   Instance Segmentation</font> 

In Instance Segmentation every pixel of object also belong to particular class but every obejct of same class don't have same(Objects with same class will have different color ).It identify each of these obeject individually 

## <font style="color:rgb(134,19,348)">   Process of Image Segmentation</font> 

Now that you are a little bit familiar with types and concept of image segmentation, now we can talk a little more in-depth The process of our code is going to be the same as we have been doing in our last lectures, we will load our CNN model weights and configuration, pre-process our data, re-configure the output and lastly we get amazing segmentations of obejects.

***But what's going on behind the scenes?*** 

Well, Instance Segmentation is formally known as RCNN Masking, RCNN stands for Region with CNN feature, meaning that RCNN is used to classify our images and segment them with some masks. RCNN doesn't use the entire picture for the input, so for instance we have image with lots of people and other object in it and we want the localize the people in that so why should we give entire image istead of giving patch or ROI where people are there. so we forward the patch towards the R-CNN and with the less time R-CNN gives us segmented cooridnates.

Now let's get into coding!

## <font style="color:rgb(134,19,348)"> Import Libraries</font>

````python
# Import Required Libraries
import numpy as np
import random
import time
import cv2
import os
import matplotlib.pyplot as plt
from tkinter import filedialog
from tkinter import Tk
````

Here we load the labels, load some colors to work with, and we also load my ```np.random.seed(50)``` we do that already, so when you code we both would have the same random see, even though they are random, we both would get the same random values.

````python
# Load the COCO class labels on which our Mask R-CNN is trained
CLASSES = open("Media/M4/Model/object_detection_classes_coco.txt").read().strip().split('\n')

# Load the set of colors that will be used when visualizing a given instance segmentation
# We use the same seed so that both our codes match even for a rand func
np.random.seed(50)

# Randomly select the colors that shall segment our image
COLORS = np.random.randint(100, 255, size=(10, 3))

# Analyzing our defined data
print("First Three Labels:",CLASSES[:3])
print("First Three Colors: \n",COLORS[:3])
print("Length of all Labels:",len(CLASSES))
````

## <font style="color:rgb(134,19,348)"> Read Image</font>


````python
# Load the input image
image = cv2.imread('Media/M4/cow.jpeg')

# Display image
plt.figure(figsize=(10,10))        
plt.imshow(image[:,:,::-1])
plt.axis('off');
````

## <font style="color:rgb(134,19,348)"> Loading the RCNN</font>

Here we load the model weights and configuration.In object dection i used **cv2.dnn.readNetFromCaffee()** but here i will use **cv2.dnn.readNetFromTensorflow()**  we put wieghts which are pre-defined with **coco classes** and configuation of dataset and weight 

````python
# Define the paths for the Mask R-CNN weights and model configuration
weights = 'Media/M4/Model/frozen_inference_graph.pb'
config = 'Media/M4/Model/mask_rcnn_inception_v2_coco_2018_01_28.pbtxt'

# Load our Mask R-CNN trained on the COCO dataset (90 classes) from disk
net= cv2.dnn.readNetFromTensorflow(weights, config)
net.setPreferableTarget(cv2.dnn.DNN_TARGET_OPENCL)
print("Done")
````

## <font style="color:rgb(134,19,348)">  Pre-processing the image 
</font>

Before inserting in dnn network we firstly need to preprocess the image because dnn requires fixed spatial dimensions and this could be acheived by one line of code.

 [```blobFromImage(image[, scalefactor[, size[, mean[, swapRB[, crop]]]]])```]( https://www.pyimagesearch.com/2017/11/06/deep-learning-opencvs-blobfromimage-works/)

**Params:**
  1. **Image**  Input image.
  2. **swapRB**  `True` Here we are swaping the RGB channels
  3. **crop**  `False` Here we won't crop our image that's why we false the crop option 

````python
# Fetching width and height if image 
(h, w) = image.shape[:2]

# Here we get the processed image
blob = cv2.dnn.blobFromImage(image, swapRB=True, crop=False)
````

## <font style="color:rgb(134,19,348)"> Input blob image to the network </font>

````python
# Passing the blob as input through the network 
net.setInput(blob)
````

## <font style="color:rgb(134,19,348)">  Forwarding Image in our Network
</font>

Here we are forwarding the input image and and in the output we get 2 values boxes and masks of image.

````python
start = time.time()
(Rect_boxes, Segmented_masks) = net.forward(["detection_out_final", "detection_masks"])
end = time.time()     
#Time it take for finding detection
print("Time it takes {:.2f} seconds".format(end - start))
````

## <font style="color:rgb(134,19,348)">  Drawing the Boxes around its detection 
</font>

We checked its out now lets verify it by drawing box around its detection. here we are going to findout that it detect correct or not

````python
# Create new copy of our image
img_copy = image.copy()

cols = 2
rows = 2

# Define Figure Size
plt.figure(figsize=(20,20))

# Iterating the detection we got
for idx,i in enumerate(range(0, Rect_boxes.shape[2]),1):
    
    # Fetching the confidence of class class label
    confidence = Rect_boxes[0, 0, i, 2]
    
    # Fetching the class index
    index = int(Rect_boxes[0, 0, i, 1])
    
    if confidence > 0.5:  
        
        # Fetching coordinates of detections
        Rect_box = Rect_boxes[0, 0, i, 3:7] * np.array([w, h, w, h])
        (X1, X2, Y1, Y2) = Rect_box.astype("int")
        
        # Display the class index and confidence score
        label = "{}: {:.2f}%".format(CLASSES[index], confidence * 100)
        
        print(label)
        
        # Here we are randomizing the color 
        color = random.randint(100,255),random.randint(200,255),random.randint(180,255)
        
        # Draw Rectangle around its suroundigs 
        cv2.rectangle(img_copy, (X1, X2), (Y1, Y2),color, 4)
        
         # Put the Label of the detected object 
        cv2.putText(img_copy, label, (X1, X2-10),cv2.FONT_ITALIC, 0.7, color, 2)
        
        # Display detections
        plt.subplot(rows,cols,idx);
        plt.imshow(img_copy[:,:,::-1])
        plt.axis('off')                
````

## <font style="color:rgb(134,19,348)">  Crop the Segment and Display </font>
In this code we are showing `Region of Interest,` `Segmentation of object` and lastly `colored segment` 

Image segmentation requires that we find all pixels where an object is present. Thus, were going to place a transparent 
overlay on top of the object to see how well our algorithm is performing. In order to do so, well calculate a mask

````python
 Rect_boxes[0, 0, 0, 3:7]
````

````python
# Here we iterate over all the objects detected in our image
for i in range(0, Rect_boxes.shape[2]):
    
    # Fetching the confidence of class class label
    confidence = Rect_boxes[0, 0, i, 2]
    
    # Fetching the class index
    index = int(Rect_boxes[0, 0, i, 1])
    
    if confidence > 0.5:
        
        # Creating copy of our original image
        img_copy = image.copy()

        # Fetching coordinates of detections
        Rect_box = Rect_boxes[0, 0, i, 3:7] * np.array([w, h, w, h])
        (X1, X2, Y1, Y2) = Rect_box.astype("int")
        
        # Fetching box width and height
        bwidth = Y1 - X1
        bheight = Y2 - X2
        
        
        
        # Here we, take out the pixel-wise segmentation for the object
        P_mask = Segmented_masks[i, index]
        
        #  Here we are resizing the mask such that it has same dimensions of the bounding box
        P_mask = cv2.resize(P_mask, (bwidth, bheight),interpolation=cv2.INTER_NEAREST) 
        
        # Here we are setting up threshold value of mask
        mask = (P_mask > 0.3)
        
        
      
        # Lets fetch the region of interest from our image
        ROI = img_copy[X2:Y2, X1:Y1]
        
        # Create copy of roi for display without any conflict
        ROI_copy=ROI.copy()
        
        # change the data type of mask and mutiplying all mask value with 255
        visual_Mask = (mask * 255).astype("uint8")
    
        # Turning the boolean mask to an integer mask
        instance = cv2.bitwise_and(ROI, ROI, mask=visual_Mask)


        # Here we are extracting the masked region of interest 
        ROI = ROI[mask]

        
        # Pick up the random color from range of 0 - 256 for every segmented object
        c = random.choice(COLORS)
        
        # Overlay by blending the randomly selected color with the ROI
        mixed = ((0.5 * c) + (0.4 * ROI)).astype("uint8")

        # Store the blended ROI in the original image
        img_copy[X2:Y2, X1:Y1][mask] = mixed
        
        
        # Draw the bounding box of the instance on the image
        color = [int(c1) for c1 in c]
        
        # Create rectangle box surround the mask
        cv2.rectangle(img_copy, (X1, X2), (Y1, Y2), color, 2)

        # Write the predicted label with its confidence 
        label = "{}: {:.3f}".format(CLASSES[index], confidence)
        cv2.putText(img_copy, label, (X1, X2 - 10),cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
        
        # Resizing them in order to show 3 pciture together
        img_copy = cv2.resize(img_copy,(ROI_copy.shape[1],ROI_copy.shape[0]))
        
        # Horizontal stack 3 picture for good aligment
        frame1=np.hstack((ROI_copy,instance,img_copy))
        
        # show the output image       
        plt.figure(figsize=(20,10)) 
        plt.imshow(frame1[:,:,::-1])
        plt.axis('off')
        plt.show()
            
                       
# Destroy all Windows            
cv2.destroyAllWindows()           
````

## <font style="color:rgb(134,19,348)"> Creating Functions </font> 
Now that we have understood step by step how to create the pipeline for classification using OpenCV's DNN module, we'll now create functions that do all the above in a single step.

### <font style="color:rgb(134,19,348)"> Initialization Function </font>
This method will be run once and it will initialize the network with the required files.

````python
def init_segmentation(usegpu = 'None', weights_name = 'frozen_inference_graph.pb',architecture_name = 'mask_rcnn_inception_v2_coco_2018_01_28.pbtxt'):
    
     # Set global variables                  
    global net,CLASSES,COLORS
    
    base_path = 'Media/M4/Model'                  
    
    # Read the Classes                  
    CLASSES = open(os.path.join(base_path,"object_detection_classes_coco.txt")).read().strip().split('\n')
    
    np.random.seed(50)
    COLORS = np.random.randint(0, 255, size=(10, 3))
    
    weights = os.path.join(base_path, weights_name)
    config = os.path.join(base_path, architecture_name)
                      
    # Intialize the model
    net= cv2.dnn.readNetFromTensorflow(weights, config)
    net.setPreferableTarget(cv2.dnn.DNN_TARGET_OPENCL)
    
     # If specified use either cuda based Nvidia gpu or opencl based Intel gpu.   
    if usegpu == 'cuda':
        net.setPreferableBackend(cv2.dnn.DNN_BACKEND_CUDA)
        net.setPreferableTarget(cv2.dnn.DNN_TARGET_CUDA)
        
    elif usegpu == 'opencl':
        net.setPreferableTarget(cv2.dnn.DNN_TARGET_OPENCL)
````

### <font style="color:rgb(134,19,348)">  Main Method </font>

````python
def segmentation(image=[], usercam=False,returndata=False, size=0.7,conf=0.5):
    
     # If the user did'nt specified the image then consider then consider choosing file or camera snapshot.
    if np.size(image) == 0:
        
        # If usercam cam is True then take picture from camera
        if usercam:
            cap=cv2.VideoCapture(0, cv2.CAP_DSHOW)
            ret, frame=cap.read()
            if ret:
                image= cv2.flip(frame,1)
                cap.release()
            else:
                print('Cant Open Camera')
                sys.exit()
        
        # Otherwise select image from computer using tkinter
        else:
            root = Tk()
            filename =  filedialog.askopenfilename(title = "Select Image")
            root.withdraw()
            image = cv2.imread(filename)
    
    
    
    img_copy = image.copy()
    h, w = img_copy.shape[:2]
    
    blob = cv2.dnn.blobFromImage(image, swapRB=True, crop=False)
    net.setInput(blob)
    
    
    # Get the detection and obejcts masks
    (Rect_boxes, Segmented_masks) = net.forward(["detection_out_final", "detection_masks"])
    
    class_data=[]
    
     # Loop over the detections
    for i in range(0, Rect_boxes.shape[2]):
        
        # Extract the confidence probablity associated with the prediction 
        confidence = Rect_boxes[0, 0, i, 2]
        
         # Extracting the index of the class.
        index = int(Rect_boxes[0, 0, i, 1])

        # Filter out detection which are less than given threshold
        if confidence > 0.5:
            
            # detected boxex coordinates
            Rect_box = Rect_boxes[0, 0, i, 3:7] * np.array([w, h, w, h])
            (X1, X2, Y1, Y2) = Rect_box.astype("int")
            
            # Calculating width and height
            bwidth = Y1 - X1
            bheight = Y2 - X2
            
            # Get the segmented maks
            P_mask = Segmented_masks[i, index]
            P_mask = cv2.resize(P_mask, (bwidth, bheight),interpolation=cv2.INTER_NEAREST) 
            mask = (P_mask > 0.3)
            
            # Get the Region of interest with its cooridnates 
            ROI = img_copy[X2:Y2, X1:Y1]
            ROI_copy=ROI.copy()
            
            # Get mask of the object 
            visual_Mask = (mask * 255).astype("uint8")           
            instance = cv2.bitwise_and(ROI, ROI, mask=visual_Mask)
            ROI = ROI[mask]
            
            # Choose the random color for mask object
            c = random.choice(COLORS)
            mixed = ((0.5 * c) + (0.4 * ROI)).astype("uint8")
            img_copy[X2:Y2, X1:Y1][mask] = mixed
            
            # Draw rectangle box and class label 
            color = [int(c1) for c1 in c]
            cv2.rectangle(img_copy, (X1, X2), (Y1, Y2), color, 2)
            label = "{}: {:.3f}".format(CLASSES[index], confidence)
            cv2.putText(img_copy, label, (X1, X2 - 10),cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
            
            img_copy = cv2.resize(img_copy,(ROI_copy.shape[1],ROI_copy.shape[0]))

            frame1=np.hstack((ROI_copy,instance,img_copy))
            frame1 = cv2.resize(frame1,(800,500))
            
            cv2.imshow("Image", frame1)
            cv2.waitKey(0)
            
            class_data.append((frame1,confidence))              
            
    if  returndata:
        return image,class_data
    else:
        cv2.destroyAllWindows()
````
