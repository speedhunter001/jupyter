##  <font style="color:rgb(134,19,348)">Canny Edge Detection</font> 
Here we are finding canny edge detection inorder to compare with Deep Edge detection

```python
# Firstly converting image into BGR to Gray Color
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Here we are applying Guassian blur 
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Finally find canny edge detection
canny = cv2.Canny(blurred, 30, 150)
```

##  <font style="color:rgb(134,19,348)"> Pre-processing the image 
</font>

```python
# Fetch image width and height
height,width = image.shape[:2]

# Prepare the frame to be fed to the network
inpBlob = cv2.dnn.blobFromImage(image,1.0,(width, height),(104.00698793, 116.66876762, 122.67891434),swapRB=False, crop=False)
```

##  <font style="color:rgb(134,19,348)">Forward Pass</font> 


```python
%%time
net.setInput(inpBlob)
output = net.forward()
```

```python
print(output.shape)
```

Above you can see we got 4 values lets understand about them a little bit.

**output:**

- **`1:`**	-( dnt know exactly ) 
- **`1:`**	 -( dnt know exactly )
- **`640:`**	 Height of the output Edge map.
- **`960:`**	 Width of the output Edge map.

```python
# Here we are resizing our final ouput to the size of input image
output = cv2.resize(output[0, 0], (width, height))

# Here we are multiplying 255 with output inorder to scale our image to 0 - 255 and change its data type to unsigned integer
output = (255 * output).astype("uint8")
```

```python
# Display the image and compare the result of canny edge with deep edge detection
plt.figure(figsize=(20,20))
plt.subplot(1,3,1);plt.imshow(image[:,:,::-1]);plt.title("Original Image");plt.axis("off");
plt.subplot(1,3,2);plt.imshow(canny);plt.title("Canny Edge Detection");plt.axis("off");
plt.subplot(1,3,3);plt.imshow(output);plt.title("Deep Edge Detection");plt.axis("off");
```

## <font style="color:rgb(134,19,348)"> Creating Functions </font> 
Now that we have understood step by step how to create the pipeline for Deep Edge Detection using OpenCV's DNN module, we'll now create functions that do all the above in a single step.

### <font style="color:rgb(134,19,348)"> Initialization Function </font>
This method will be run once and it will initialize the network with the required files.

```python
def init_edgedetection(usegpu='None',weights_name="hed_pretrained_bsds.caffemodel",proto_name="deploy.prototxt",):
    
    # Set global variables
    global net
    
    base_path = 'Media/M4/Model'
    
    # Defining model paths
    proto_file = os.path.join(base_path, proto_name)
    weights = os.path.join(base_path, weights_name) 
    
    # Initialize the DNN module
    net = cv2.dnn.readNetFromCaffe(proto_file, weights)
    
     # If specified use either cuda based Nvidia gpu or opencl based Intel gpu.   
    if usegpu == 'cuda':
        net.setPreferableBackend(cv2.dnn.DNN_BACKEND_CUDA)
        net.setPreferableTarget(cv2.dnn.DNN_TARGET_CUDA)
        
    elif usegpu == 'opencl':
        net.setPreferableTarget(cv2.dnn.DNN_TARGET_OPENCL)
```

### <font style="color:rgb(134,19,348)">  Main Method </font>

```python
def edgedetection(image=[], usercam=False,returndata=False):
    
    # If the user did'nt specified the image then consider then consider choosing file or camera snapshot.
    if np.size(image) == 0:
        
        # If usercam cam is True then take picture from camera
        if usercam:
            cap=cv2.VideoCapture(0, cv2.CAP_DSHOW)
            ret, frame=cap.read()
            if ret:
                image= cv2.flip(frame,1)
                cap.release()
            else:
                print('Cant Open Camera')
                sys.exit()
                
        # Otherwise select image from computer using tkinter        
        else:
            root = Tk()
            root.filename =  filedialog.askopenfilename(title = "Pick up the Image")
            root.withdraw()
            image = cv2.imread(root.filename)
            
    (height,width) = image.shape[:2]
    
    # Prepare the frame to be fed to the network
    inpBlob = cv2.dnn.blobFromImage(image,1.0,(width, height),(104.00698793, 116.66876762, 122.67891434),
                                    swapRB=False, crop=False)
    
    # Forward pass
    net.setInput(inpBlob)
    output = net.forward()
    
    # Here we are resizing our final ouput to the size of input image
    output = cv2.resize(output[0, 0], (width, height))

    # Here we are multiplying 255 with output inorder to scale our image to 0 - 255 and change its data type to unsigned integer
    output = (255 * output).astype("uint8")

    
            
    if  returndata:
        return output
    else:
        plt.figure(figsize=(15,15))
        plt.subplot(1,2,1);plt.imshow(image[:,:,::-1]);plt.title("Original Image");plt.axis("off");
        plt.subplot(1,2,2);plt.imshow(output);plt.title("Deep Edge Detection");plt.axis("off");          
```

## <font style="color:rgb(134,19,348)"> Real time Deep Edge Detection </font>

```python
fps=0
init_edgedetection()
cap=cv2.VideoCapture(0)

while(True):    
    
    start_time = time.time()
    ret,frame=cap.read() 
    image=cv2.flip(frame,1)
    
    image = edgedetection(image,returndata=True)
    
    cv2.putText(image, 'FPS: {:.2f}'.format(fps), (400, 20), cv2.FONT_HERSHEY_SIMPLEX,0.8, (255, 20, 55), 2)
    cv2.imshow("Edge detection",image)
    
    k = cv2.waitKey(1)
    fps= (1.0 / (time.time() - start_time))
    
    if k == ord('q'):
        break
cap.release() 
cv2.destroyAllWindows() 
```
