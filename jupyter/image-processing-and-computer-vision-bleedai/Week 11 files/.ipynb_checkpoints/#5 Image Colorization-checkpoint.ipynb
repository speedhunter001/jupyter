#  <center> <font style="color:rgb(134,19,348)"> Image Colorization </font> </center>
Basically Image Colorization is the art of converting `Grayscale Image`(Input) to the rich colored`Colorized Image`( Output). Traditional method of image colorization is so much hectic and required human input to manually set the color on image objects and takes too much time for conversion.But in this notebook we are going to use modern techniques of colorization which requries no human input during the time of conversion and convert the image with in seconds.


## <font style="color:rgb(134,19,348)"> What is structure behind the conversion of image </font>

<center><img src="Media/M4/colorization_arch.PNG"/></center>

**Credits:**  `Zhang et al.â€™s architecture for colorization of black and white images with deep learning.`

This technique is published by Zhang et al. he take start from ImageNet Dataset and convert all RGB image to the Lab color space. basically Lab color space is same as RGB and consist of 3 values, but only the difference b/w RGB and Lab color space is that in Lab **color information Encode differently**. 

Lets's beak out Lab and understand in bit detail
  - **L**  L channel encodes lightness intensity
  - **a** In a channel green and red color encoded.
  - **b** In b channel blue and yellow color encoded.
  
So now onwards for the input(Grayscale image) we are only concerned with L channel which is not concern with any color channel and we only give input L channel and in the output we predict a,b channel which are concern with color and finally we add colored channel in our Grayscale image and get the colored image.

**Complete Process**
- `Convert Grayscale image to Lab color space`
- `Fetch the L channel from Lab color`
- `Give L channel as input to the netowork and predict the a,b channel `
- `Than combine origianl L channel with predicted ab channels`
- `Lastly convert the Lab image to RGB`



```python
# Import required libraries
import numpy as np
import cv2
import time
import matplotlib.pyplot as plt
from tkinter import filedialog
from tkinter import Tk
import sys
import os
```

## <font style="color:rgb(134,19,348)"> Initialize DNN Module</font>

```python
# Load the Architecture.
architecture = 'Media/M4/Model/colorization_deploy_v2.prototxt'
# Load the Model Weights.
weights = 'Media/M4/Model/colorization_release_v2.caffemodel'
```

```python
# Here we are reading colorization model with caffe
net = cv2.dnn.readNetFromCaffe(architecture, weights)
print("Model loaded Succesfully")
```

## <font style="color:rgb(134,19,348)"> Load the cluster center points </font>


```python
# Cluster center point file is in numpy format that's why we are loading with numpy
points_path = 'Media/M4/Model/pts_in_hull.npy'
points = np.load(points_path)
print(points.shape)
```

## <font style="color:rgb(134,19,348)"> Read the Image </font>

```python
# Load the input image from disk
image = cv2.imread('Media/M4/old2.jpeg')

# Display it
plt.figure(figsize=(10,10))
plt.imshow(image[:,:,::-1])
plt.axis("off");
```

## <font style="color:rgb(134,19,348)"> Get a,b channels </font>


```python
# Here we are fetching ab channel from our network
cls_8 = net.getLayerId("class8_ab")
conv_8 = net.getLayerId("conv8_313_rh")
print(cls_8, conv_8)
```

## <font style="color:rgb(134,19,348)"> Transpose the points </font>


```python
points = points.transpose()
points = points.reshape(2, 313, 1, 1)
print(points.shape)
```

## <font style="color:rgb(134,19,348)"> Load Center for a,b channel </font>


```python
# Here we are loading center for ab channel and normalize it for adjustment
points_float32 =  [points.astype("float32")]
net.getLayer(cls_8).blobs = points_float32

points_full =   [np.full([1, 313], 2.606, dtype="float32")]
net.getLayer(conv_8).blobs = points_full
```

```python
# Changing the data type to float32 and divide it 255 so the value should be b/w 0 - 1
scale_pixl = image.astype("float32") / 255.0
print(scale_pixl.shape)
```

## <font style="color:rgb(134,19,348)"> Convert BGR TO Lab </font>


```python
# Here we are converting the BRG image into lab color space
lab_colr_space = cv2.cvtColor(scale_pixl, cv2.COLOR_BGR2LAB)

# Colorization network accpet a specific dimension so we resize it in required dimensions
lab224 = cv2.resize(lab_colr_space, (224, 224))
```

## <font style="color:rgb(134,19,348)">  Fetch the L channels </font>


```python
# Fetching the L channel from resized lab color space
l_channel = lab224[:,:,0] # pull out L channel

# Mean subtraction 
l_channel -= 50  
```

## <font style="color:rgb(134,19,348)">  Input L channel as input in Network </font>


```python
# So here we are passing L channel in the network which will predict a,b channel
net.setInput(cv2.dnn.blobFromImage(l_channel))
```

```python
%%time
# Predicted a,b channel 
ab_channel = net.forward()
```

```python
# Traspose a,b channel
print(ab_channel.shape) 
ab_channel = ab_channel[0, :, :, :].transpose((1, 2, 0))
print(ab_channel.shape)
```

```python
# Here we are resizing a,b channel with simillar dimension of our original image
a_b = cv2.resize(ab_channel, (image.shape[1], image.shape[0]))
```

## <font style="color:rgb(134,19,348)"> Verifying the L and ab channel </font>


```python
# Here we are fetching L chanel from original image
l_channel = lab_colr_space[:,:,0]
print("shape of L Channel"+str(l_channel.shape))
print("shape of ab Channel"+str(a_b.shape))
```

## <font style="color:rgb(134,19,348)">  Merge L and ab channel </font>


```python
# Concatenate both L channel from  original image and ab channel which are predicted
colored = np.concatenate((l_channel[:, :, np.newaxis], a_b), axis=2)
print(colored.shape)
```

## <font style="color:rgb(134,19,348)">  Convert to LAB to RGB  </font>

```python
# Now again converting the lab color to RGB channel so we can visualize the image 
# clip any values that fall outside the range [0, 1]
colored = cv2.cvtColor(colored, cv2.COLOR_LAB2BGR)

# Firstly we clip the value b/w 0 and 1 range now we are clipping values that excludes 0 and 1 
colored = np.clip(colored, 0, 1)
```

```python
# Multiplying to 255 so all float value become integer and changing data type to unsigned 8 bit interger
colored = (255 * colored).astype("uint8")
```

```python
# Stack original and colorized image together
Final = np.hstack((image,colored))

# Display the final image
plt.figure(figsize=(10,10))
plt.imshow(Final[:,:,::-1])
plt.text(400,-20,"Original Image")
plt.text(1500,-20,"Colorized Image")
plt.axis("off");
```

## <font style="color:rgb(134,19,348)"> Creating Functions </font> 
Now that we have understood step by step how to create the pipeline for classification using OpenCV's DNN module, we'll now create functions that do all the above in a single step.

### <font style="color:rgb(134,19,348)"> Initialization Function </font>
This method will be run once and it will initialize the network with the required files.

```python
def init_colorized(usegpu = 'None', weights_name = 'colorization_deploy_v2.prototxt',architecture_name ='colorization_release_v2.caffemodel'):
    
    # Set global variables
    global net,points
    
    base_path = 'Media/M4/Model'
    
    protofile = os.path.join(base_path, weights_name);
    model = os.path.join(base_path, architecture_name)
    
    # Load the points
    points_path = os.path.join(base_path,'pts_in_hull.npy')
    points = np.load(points_path)
    
     # Intialize the model
    net = cv2.dnn.readNetFromCaffe(protofile, model)
    
    cls_8 = net.getLayerId("class8_ab")
    conv_8 = net.getLayerId("conv8_313_rh")
    
    points = points.transpose().reshape(2, 313, 1, 1)
    
    net.getLayer(cls_8).blobs = [points.astype("float32")]
    net.getLayer(conv_8).blobs = [np.full([1, 313], 2.606, dtype="float32")]
    
    # If specified use either cuda based Nvidia gpu or opencl based Intel gpu.   
    if usegpu == 'cuda':
        net.setPreferableBackend(cv2.dnn.DNN_BACKEND_CUDA)
        net.setPreferableTarget(cv2.dnn.DNN_TARGET_CUDA)
        
    elif usegpu == 'opencl':
        net.setPreferableTarget(cv2.dnn.DNN_TARGET_OPENCL)
```

### <font style="color:rgb(134,19,348)"> Main Method</font>


```python
def colorization(image=[],returndata=False):
    
    # If the user did'nt specified the image then consider then consider choosing file or camera snapshot.
    if np.size(image) == 0:
        
         # Ohterwise Select image from computer using tkinter
        root = Tk()
        root.filename =  filedialog.askopenfilename(title = "Pick up the Image")
        root.withdraw()
        image = cv2.imread(root.filename)
        
    
    scale_pixl = image.astype("float32") / 255.0
    
    lab_colr_space = cv2.cvtColor(scale_pixl, cv2.COLOR_BGR2LAB)
    lab224 = cv2.resize(lab_colr_space, (224, 224))
    
    # Pull out L channel
    l_channel = lab224[:,:,0] 
    l_channel -= 50  
    
    net.setInput(cv2.dnn.blobFromImage(l_channel))
    ab_channel = net.forward()
    
    ab_channel = ab_channel[0, :, :, :].transpose((1, 2, 0))
    a_b = cv2.resize(ab_channel, (image.shape[1], image.shape[0]))
    
    l_channel = lab_colr_space[:,:,0]
    
    # Colorized the image
    colorized = np.concatenate((l_channel[:, :, np.newaxis], a_b), axis=2)
    colorized = cv2.cvtColor(colorized, cv2.COLOR_LAB2BGR)
    colorized = np.clip(colorized, 0, 1)
    colorized = (255 * colorized).astype("uint8")
    
    
    Final = np.hstack((image,colorized))
    #Final = cv2.resize(Final,(700,500))
    
    if  returndata:
        return Final
    else:
        # Display the final image
        plt.figure(figsize=(10,10))
        plt.imshow(Final[:,:,::-1])
        plt.text(400,-20,"Original Image")
        plt.text(1500,-20,"Colorized Image")
        plt.axis("off");          
```

```python
init_colorized()
```

```python
colorization()
```

## <font style="color:rgb(134,19,348)">  Video Colorization</font>

```python
# Initializing method
init_colorized()

# Starting video reading mode
cap = cv2.VideoCapture('Media/M4/oldy.mp4')
fourcc = cv2.VideoWriter_fourcc(*'XVID')

# Specifying Video writing path
out = cv2.VideoWriter(r'Media/M4/colorized.mp4',fourcc, 20.0, (700,500))

while(True):
    
    ret, frame = cap.read()
    if not ret:
        break
        
    frame = colorization(frame,returndata=True)
    
    cv2.imshow('Orignal VS Colorized Frame',frame)
    out.write(frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    #print(final.shape)
cap.release()
out.release()
cv2.destroyAllWindows()
```
