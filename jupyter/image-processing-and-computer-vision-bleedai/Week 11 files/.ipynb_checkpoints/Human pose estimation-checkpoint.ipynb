# <center> <font style="color:rgb(100,109,254)"> Human Pose Estimation with OpenPose</font> </center>

**What is Human pose estimation?**

human pose estimation can be defined as the computer vision techniques that predict the location of various human keypoints(joints and landmarks) such as elbows, knees, neck, shoulder.sometime it becomes very difficult problem because of illumination variablity and changes in body posture.

## <center> <font style="color:rgb(100,109,254)"> What is OpenPose</font> </center>
The openPose project from Carnegie Mellon University is a well know method and most accurate method for pose estimation using CNN.for finding pose it uses multistage classifier where each stage improves result from its previous.

**Basically there are three steps to acheive it:**

`Keypoints localization` First step is is responsible for predicting all the keypoints with a confidence score. This is called confidence Map.

`Part Affinity Fields ` In Second step the network predicts a 2D vector field that predicts the keypoints association data.
 
 `Greedy Inference` Finally, we connect all the keypoints using greedy inference.
 
 

## <font style="color:rgb(100,109,254)">Import Libraries</font> 


```python
# Import Required libraries
import cv2
import time
import numpy as np
import matplotlib.pyplot as plt
from tkinter import filedialog
from tkinter import Tk
%matplotlib inline
```

## <font style="color:rgb(100,109,254)">Define the Points</font> 


```python
# Here we are defining the each point for each body part
p = {"Head": 0, "Neck": 1, "RShoulder": 2, "RElbow": 3, "RWrist": 4,
"LShoulder": 5, "LElbow": 6, "LWrist": 7, "RHip": 8, "RKnee": 9,
"RAnkle": 10, "LHip": 11, "LKnee": 12, "LAnkle": 13, "Chest": 14,
"Background": 15}

# Arranging the points according to its pre-defined structure 
pose_sets=[ [p["Head"], p["Neck"]],  [p["Neck"], p["RShoulder"]], [p["RShoulder"], p["RElbow"]],
             [p["RElbow"], p["RWrist"]], [p["Neck"], p["LShoulder"]], [p["LShoulder"], p["LElbow"]],
             [p["LElbow"], p["LWrist"]], [p["Neck"], p["Chest"]], [p["Chest"], p["RHip"]], [p["RHip"], p["RKnee"]],
             [p["RKnee"], p["RAnkle"]], [p["Chest"], p["LHip"]], [p["LHip"], p["LKnee"]], [p["LKnee"], p["LAnkle"]]] 
```

## <font style="color:rgb(100,109,254)">Loading the Model</font> 


```python
# Fetching paths of models 
proto_file = "Model/pose_deploy_linevec_faster_4_stages.prototxt"
weights_file = "Model/pose_iter_160000.caffemodel"
 
# Read the wieght and model file  
net = cv2.dnn.readNetFromCaffe(proto_file, weights_file)
```

## <font style="color:rgb(100,109,254)">Read Image</font> 


```python
# Read image
image = cv2.imread("Media/single2.jpg")

# Display image
plt.figure(figsize=[10,10])
plt.imshow(image[:,:,::-1])
plt.axis('off');
```

```python
width = image.shape[1]
height = image.shape[0]
width,height
```

## <font style="color:rgb(100,109,254)"> Pre-processing the image 
</font>

```python
# Prepare the frame to be fed to the network
inpBlob  = cv2.dnn.blobFromImage(image, 1.0 / 255, (300, 300), (0, 0, 0), swapRB=False, crop=False)
```

## <font style="color:rgb(100,109,254)">Fetching Pose Detections</font> 


```python
net.setInput(inpBlob)
output = net.forward()
```

```python
print(output.shape)
```

Above you can see we got 4D matrix lets understand about them a little bit.

**output:**

- **`1:`**	ImageID for one image if you pass more than 1 image it vaies 
- **`44:`**	 for MPI, it produces 44 points that indicates the index of a keypoint in confidence map
- **`38:`**	 Height of resultant image 
- **`38:`**	 Width of resultant  iamge 

```python
H = output.shape[2]
W = output.shape[3]
```

## <font style="color:rgb(100,109,254)">Iterate all points with probablity map </font> 


```python
# Defining rows and columns
rows = 4;
cols=4;

# Define Figure Size
plt.figure(figsize=(20,20));

# Show all points with prob map
for index, i in enumerate(p, 1):
    
    # Fetch the key point
    kp =p[i]
    
    # Fetch the specific point from porbablity map 
    pro_map = output[0, kp, :, :]
    
    # Here we have create the probablity same of image size 
    pro_map = cv2.resize(pro_map, (width, height))
    
    # Display it with probablity map 
    plt.subplot(rows,cols,index);plt.imshow(image[:,:,::-1]);plt.axis('off');

    # If you remove the probablity map you can't see the point in image
    plt.imshow(pro_map, alpha=0.5); plt.title("{} point".format(i));plt.axis('off');  
```

## <font style="color:rgb(100,109,254)">Draw All individual points on original image </font> 


```python
# Creating copy of image for showing point and skeleton of points 
img_copy = image.copy()
points = []

# Defining nummber of point we want iterate on image
number_of_points = 15

# Threshold value for the points to be consider 
threshold = 0.1

for i in range(number_of_points):
    
    # Fetch the specific point from porbablity map 
    pro_map = output[0, i, :, :]
    
    # Find minimum and maximum value of probablity map 
    _, probablity, _, point = cv2.minMaxLoc(pro_map)
        
    # Resize the points inorder to fit on image 
    x = (width * point[0]) / W
    y = (height * point[1]) / H
    
    # If probablity value above than given threshold than draw consider that point
    if probablity > threshold :
      
        # Draw circle and put number on each point 
        cv2.circle(img_copy, (int(x), int(y)), 8, (255, 0, 255), thickness=-1, lineType=cv2.FILLED)
        cv2.putText(img_copy, "{}".format(i), (int(x), int(y)), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, lineType=cv2.LINE_AA)

        # Append point coordinates for futher use
        points.append((int(x), int(y)))
    else :
        points.append(None)

# Display image   
plt.figure(figsize=[10,10])        
plt.imshow(img_copy[:,:,::-1])
plt.title("Person With all detected points ")        
plt.axis('off');
```

## <font style="color:rgb(100,109,254)">Draw Skeleton </font> 


```python
img_copy1 = image.copy()

for i in range(number_of_points):
   
  
    # Fetch the specific point from porbablity map 
    pro_map = output[0, i, :, :]
    
    # Find minimum and maximum value of probablity map 
    _, probablity, _, point = cv2.minMaxLoc(pro_map)
    
    
    #  Resize the points inorder to fit on image 
    x = (width * point[0]) / W
    y = (height * point[1]) / H

    if probablity > threshold : 
        cv2.circle(img_copy1, (int(x), int(y)), 8, (255, 255, 0), thickness=-1, lineType=cv2.FILLED)

        # Add the point to the list if the probability is greater than the threshold
        points.append((int(x), int(y)))
    else :
        points.append(None)

# Tracing point according to its sekeleton and make skeleton of body
for pair in pose_sets:
    A1 = pair[0]
    B1 = pair[1]

    if points[A1] and points[B1]:
        cv2.line(img_copy1, points[A1], points[B1], (0, 0, 255), 3)

# Display image         
plt.figure(figsize=[10,10])
plt.imshow(img_copy1[:,:,::-1])
plt.title("Points Traced with its pair")        
plt.axis('off');
```

## <font style="color:rgb(100,109,254)">  Iniitialize the method </font>

```python
def init_pose():
    global p,pose_sets,net

    pose = {"Head": 0, "Neck": 1, "RShoulder": 2, "RElbow": 3, "RWrist": 4,
    "LShoulder": 5, "LElbow": 6, "LWrist": 7, "RHip": 8, "RKnee": 9,
    "RAnkle": 10, "LHip": 11, "LKnee": 12, "LAnkle": 13, "Chest": 14,
    "Background": 15}

    pose_sets=[ [pose["Head"], pose["Neck"]],  [pose["Neck"], pose["RShoulder"]], [pose["RShoulder"], pose["RElbow"]],
                 [pose["RElbow"], pose["RWrist"]], [pose["Neck"], pose["LShoulder"]], [pose["LShoulder"], pose["LElbow"]],
                 [pose["LElbow"], pose["LWrist"]], [pose["Neck"], pose["Chest"]], [pose["Chest"], pose["RHip"]], [pose["RHip"], pose["RKnee"]],
                 [pose["RKnee"], pose["RAnkle"]], [pose["Chest"], pose["LHip"]], [pose["LHip"], pose["LKnee"]], [pose["LKnee"], pose["LAnkle"]]]
    
    proto_file = "Model/pose_deploy_linevec_faster_4_stages.prototxt"
    weights_file = "Model/pose_iter_160000.caffemodel"

    net = cv2.dnn.readNetFromCaffe(proto_file, weights_file)
```

```python
def pose(image=[],usercam=False,returndata=False,threshold=0.5,size=0.7):
    
    if np.size(image) == 0:

        if usercam:
            cap=cv2.VideoCapture(0)
            ret,frame=cap.read()
            image= cv2.flip(frame,1)
            cap.release()
        else:
            root = Tk()
            root.filename =  filedialog.askopenfilename(title = "Pick up the Image")
            root.withdraw()
            image = cv2.imread(root.filename)
            
    (height, width) = image.shape[:2]
    blob = cv2.dnn.blobFromImage(image, 1.0 / 255, (300, 300), (0, 0, 0), swapRB=False, crop=False)
    points = []
    net.setInput(blob)
    output = net.forward()
    
    H = output.shape[2]
    W = output.shape[3]
    
    
    for i in range(15):

        pro_map = output[0, i, :, :]

        _,probablity, _, point = cv2.minMaxLoc(pro_map)

        x = (width * point[0]) / W
        y = (height * point[1]) / H

        if probablity > threshold : 
            cv2.circle(image, (int(x), int(y)), 8, (255, 255, 0), thickness=-1, lineType=cv2.FILLED)

            points.append((int(x), int(y)))
        else :
            points.append(None)

    for pair in pose_sets:
        A1 = pair[0]
        B1 = pair[1]

        if points[A1] and points[B1]:
            cv2.line(image, points[A1], points[B1], (0, 0, 255), 3)
            
    if  returndata:
        return image
    else:
        cv2.imshow("Pose Estimation", image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()          
```

## <font style="color:rgb(100,109,254)"> Real time Pose Estimation </font>

```python
fps=0
init_pose()
cap=cv2.VideoCapture(0)

while(True):    
    
    start_time = time.time()
    ret,frame=cap.read() 
    image=cv2.flip(frame,1)
    
    image = pose(image,returndata=True)
    
    cv2.putText(image, 'FPS: {:.2f}'.format(fps), (400, 20), cv2.FONT_HERSHEY_SIMPLEX,0.8, (255, 20, 55), 2)
    cv2.imshow("Pose Estimation",image)
    
    k = cv2.waitKey(1)
    fps= (1.0 / (time.time() - start_time))
    
    if k == ord('q'):
        break
cap.release() 
cv2.destroyAllWindows() 
```
