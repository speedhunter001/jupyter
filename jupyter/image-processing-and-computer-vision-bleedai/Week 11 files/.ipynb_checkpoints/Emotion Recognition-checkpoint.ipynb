```python
# Import Required libraries
import cv2
import time
import numpy as np
import matplotlib.pyplot as plt
from tkinter import filedialog
from tkinter import Tk
import os
import bleedfacedetector as fd
%matplotlib inline
```

##  <font style="color:rgb(134,19,348)">Select Model Path</font> 
select the required model path and define the emotions

```python
model = 'Media/M4/Model/emotion-ferplus-8.onnx'

# Define the emotions
emotions = ['Neutral', 'Happy', 'Surprise', 'Sad', 'Anger', 'Disgust', 'Fear', 'Contempt']
```

## <font style="color:rgb(134,19,348)"> Initialize the DNN module </font>

```python
net = cv2.dnn.readNetFromONNX(model)
```

##  <font style="color:rgb(134,19,348)">Read Image</font> 


```python
# Read image
image = cv2.imread("Media/M4/surprise.jpeg")

# Display image
plt.figure(figsize=[10,10])
plt.imshow(image[:,:,::-1])
plt.axis('off');
```

##  <font style="color:rgb(134,19,348)">Detect faces with bleedfacedetector</font> 


```python
img_copy = image.copy()
faces = fd.ssd_detect(img_copy,conf=0.2)
```

```python
# Lets take coordinates of face and understand it 
x,y,w,h = faces[0]

# Define padding for face roi
padding = 3

# Get the Face from image 
face = img_copy[y-padding:y+h+padding,x-padding:x+w+padding] 
```

##  <font style="color:rgb(134,19,348)"> Pre-processing the image 
</font>

```python
face = cv2.resize(face[:,:,1], (64, 64))

imager = face.reshape(1,1,64,64)
```

##  <font style="color:rgb(134,19,348)">Forward Pass</font> 


```python
%%time
net.setInput(imager)
Output = net.forward()
```

```python
print("Shape of the output"+str(Output.shape))
print("Predicted emotion is: "+emotions[Output.argmax()])
print("Every emotion probablity: "+str(Output))
```

```python
predicted_emotion = emotions[Output.argmax()]
```

##  <font style="color:rgb(134,19,348)">Predict the Emotion</font> 


```python
cv2.putText(img_copy,'{}'.format(predicted_emotion),(x,y+h+75), cv2.FONT_HERSHEY_SIMPLEX, 3, (255,0,255), 4, cv2.LINE_AA)
cv2.rectangle(img_copy,(x,y),(x+w,y+h),(0,0,255),5)
plt.figure(figsize=(10,10))
plt.imshow(img_copy[:,:,::-1]);plt.axis("off");
```

## <font style="color:rgb(134,19,348)"> Creating Functions </font> 
Now that we have understood step by step how to create the pipeline for Emotion Recognition using OpenCV's DNN module, we'll now create functions that do all the above in a single step.

### <font style="color:rgb(134,19,348)"> Initialization Function </font>
This method will be run once and it will initialize the network with the required files.

```python
def init_emotion(usegpu='None',model="emotion-ferplus-8.onnx"):
    
    # Set global variables
    global net
    
    base_path = 'Media/M4/Model'
    
    # Defining model path
    model = os.path.join(base_path, model)    
    
    # Initialize the DNN module
    net = cv2.dnn.readNetFromONNX(model)
    
     # If specified use either cuda based Nvidia gpu or opencl based Intel gpu.   
    if usegpu == 'cuda':
        net.setPreferableBackend(cv2.dnn.DNN_BACKEND_CUDA)
        net.setPreferableTarget(cv2.dnn.DNN_TARGET_CUDA)
        
    elif usegpu == 'opencl':
        net.setPreferableTarget(cv2.dnn.DNN_TARGET_OPENCL)
```

### <font style="color:rgb(134,19,348)">  Main Method </font>

```python
def emotion(image=[], usercam=False,returndata=False,l_thickness=2,t_size=1,t_density=2):
    
    # If the user did'nt specified the image then consider then consider choosing file or camera snapshot.
    if np.size(image) == 0:
        
        # If usercam cam is True then take picture from camera
        if usercam:
            cap=cv2.VideoCapture(0, cv2.CAP_DSHOW)
            ret, frame=cap.read()
            if ret:
                image= cv2.flip(frame,1)
                cap.release()
            else:
                print('Cant Open Camera')
                sys.exit()
                
        # Otherwise select image from computer using tkinter        
        else:
            root = Tk()
            root.filename =  filedialog.askopenfilename(title = "Pick up the Image")
            root.withdraw()
            image = cv2.imread(root.filename)
            
    # Make copy of  image
    img_copy = image.copy()
    
    # Detect face in image
    faces = fd.ssd_detect(img_copy,conf=0.2)
    
    # Lets take coordinates of face and understand it 
    x,y,w,h = faces[0]

    # Define padding for face roi
    padding = 3

    # Get the Face from image 
    face = img_copy[y-padding:y+h+padding,x-padding:x+w+padding] 

    face = cv2.resize(face[:,:,1], (64, 64))
    imager = face.reshape(1,1,64,64)
    
    net.setInput(imager)
    Output = net.forward()
    
    predicted_emotion = emotions[Output.argmax()]

    cv2.putText(img_copy,'{}'.format(predicted_emotion),(x,y+h+(t_size*20)), cv2.FONT_HERSHEY_SIMPLEX, t_size, (255,0,255), t_density, cv2.LINE_AA)
    cv2.rectangle(img_copy,(x,y),(x+w,y+h),(0,0,255),l_thickness)
    
    if  returndata:
        return img_copy
    else:
        plt.figure(figsize=(10,10))
        plt.imshow(img_copy[:,:,::-1]);plt.axis("off");          
```

## <font style="color:rgb(134,19,348)"> Real time Emotion Recognizer</font>

```python
fps=0
init_emotion()
cap=cv2.VideoCapture(0)

while(True):    
    
    start_time = time.time()
    ret,frame=cap.read() 
    image=cv2.flip(frame,1)
    
    image = emotion(image,returndata=True)
    
    cv2.putText(image, 'FPS: {:.2f}'.format(fps), (400, 20), cv2.FONT_HERSHEY_SIMPLEX,0.8, (255, 20, 55), 2)
    cv2.imshow("Emotion Recognizer",image)
    
    k = cv2.waitKey(1)
    fps= (1.0 / (time.time() - start_time))
    
    if k == ord('q'):
        break
cap.release() 
cv2.destroyAllWindows() 
```
