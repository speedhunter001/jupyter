## <font style="color:rgb(134,19,348)"> Opencv DNN Module  </font>
Opencv package includes multiple sub packages, DNN is also one of the module among them
It was introduced after opencv 3.1 and now in the latest version of opencv 4.0+ the DNN module is much imporved

The DNN module allows you to run different types of deep learning models, for e.g. models for image classification , object detection, image segmentation etc.
But One big limitation of this module is that you can train your network in DNN module you canno
t do training of any model but you can only do inference on it.
Inference implies 'Testing' which means that you pick up a pre-trained model and set in dnn module and apply inference on it lastly, get your predictions.

### <font style="color:rgb(134,19,348)"> Example of classifcation of dogs and cats with the Dnn module </font>

1. First we take a pre-trained image classifier model trained to recognize images of cats and dogs
2. Then we use this model in the dnn module.
3. After that we preprocess the image and pass it into this model.
4. Now we can do inference on this image to determine if its a cat or a dog.

````python
# Importing Required libraries
import numpy as np
import time
import cv2
import matplotlib.pyplot as plt
````

### <font style="color:rgb(134,19,348)">  Loading Classes Label </font>

Here we are reading the `synset_words.txt` ( which contain 1000 pre-trained class labels of different object like: goldfish,house,flamingo etc. ). Every class label is on one new line with its unique id and we want to store it in a list so thats why we have to split the labels by new line

````python
# Split the class variable by new line and store it in variable rows
rows = open('Model/synset_words.txt').read().strip().split("\n")

# Here we verifying length of the classlabels length
print("Length of class labels is:"+str(len(rows)))

# Printing 0 to 2 items from the list inorder to checkout the labels name
print(rows[0:2])
````

#### <font style="color:rgb(134,19,348)"> Let's undertand the first element </font>


````python
# let's perform some processing on the first element and than all the elements 
eg = rows[0]
print(eg)

# here we find out the space in this string and print all that index after space
print(eg[eg.find(" ")+1:])

# lastly we split these two string with commna 
print(eg[eg.find(" ")+1:].split(','))

# here we printed the first element which is the class name and now we apply same steps for all 
print(eg[eg.find(" ")+1:].split(',')[0])
````

### <font style="color:rgb(134,19,348)">Now for all class labels </font>

here we are applying the above 4 steps for whole 1000 class labels in which we want to extract the first elements of every class by iterating the list of class labels for element,and finding space into it adding it with 1 and split it with comma lastly extracting the first element of the class. 

````python
# Applying the above process
CLASSES = [r[r.find(" ") + 1:].split(",")[0] for r in rows]

# print the 5 procesed class labels 
print(CLASSES[0:5])
````

###  <font style="color:rgb(134,19,348)"> Read the Image </font>

````python
# load the input image
image = cv2.imread('Media/parrot.jpeg')

# Display the image
plt.imshow(image[:,:,::-1])
plt.axis("off")
````

### <font style="color:rgb(134,19,348)">  Load Our Serialized Model </font>

**Caffemodel:** It gives output at a specified interval while training,also its a binary file contains the number of pre-trained model weights.

**Prototxt:** it is generated alongside, is a binary file contains the information required to continue training the model from where it last stopped contain the architecture inorder to map the model.

````python
# There are two main file which we used in dnn network

# 1. weights 2. architecture

# This file contain the number of pre-trained model weights 
weight='Model/bvlc_googlenet.caffemodel'

# This file contain the googlenet architecture used for map the model with this architecture 
arch='Model/bvlc_googlenet.prototxt' 
````

### <font style="color:rgb(134,19,348)">  Initializing DNN Module Using Opencv </font>

Opencv dnn used with multiple models Like: Caffe, tensorflow or pytroch for that they have relatively same methods: <br>
* cv2.dnn.readNetFromCaffe                                                               

* cv2.dnn.readNetFromTensorFlow                                                                                            
* cv2.dnn.readNetFromTorch

````python
# Here we are reading pre-trained caffe model with its architecture using opencv dnn module which accepts only 2 parameters 
net = cv2.dnn.readNetFromCaffe(arch, weight)
````

### <font style="color:rgb(134,19,348)"> Pre-processing the image before inserting in network  </font>
The DNN we use requires fixed spatial dimensions, and will require our image to be preprocessed like , being normalized , mean subtracted and passed like a 4d tensor. We can do all the image processing steps manually but we have very amazing dnn function by opencv
**cv2.dnn.blobFromImage()** which will do all the preprocessing and convert our image in the required format for us.

````python
blob = cv2.dnn.blobFromImage(image, 1, (224, 224), (104, 117, 123))
````

```Python
blobFromImage(image[, scalefactor[, size[, mean[, swapRB[, crop]]]]])```

**Params:**
  1. **Image**  Input image.
  2. **Scalefactor** `1` here we set up 1 as scaling which is a default parameter.
  3. **Size** `(244,244)` the size depends upon the training of the model. Googlenet was trained for this specific size.
  4.  **Mean** `(104,117,123)` RGB values are subtracted to give illumination invariance to the model

You can read more about blobFromImage [here]( https://www.pyimagesearch.com/2017/11/06/deep-learning-opencvs-blobfromimage-works/)

### <font style="color:rgb(134,19,348)"> Input blob image to the network </font>

````python
# Passing the blob as input through the network 
net.setInput(blob)
````

###  <font style="color:rgb(134,19,348)"> Getting Predictions from the Neural Net </font>
Here the actual computation will take place, this will perform a forward pass through the network and we will get the output. 
The predictions will be the probablites of our 1000 classes, from here we can either get the class with the highest probability or  get relative higest probablity according to its train data. 
We are also using the time module to find out how much time it takes for a single pass, this is pretty useful information and it tells us if our model is capable of running in real time or not

````python
# Calculating time for prediction
start=time.time()

# Getting Prediction 
prediction = net.forward()

end=time.time()

print("classification took {:.2} seconds".format(end-start))
````

###  <font style="color:rgb(134,19,348)">Classify the object in image</font>


````python
# Resize aspect ratio of the image 
Height = 700
scale = Height / image.shape[0]
image = cv2.resize(image, None, fx=scale, fy=scale)

# Here we are applying all above techniques which we have understood one by one
idxs = np.argsort(prediction[0])[::-1][:1][0]
idxs = np.argmax(prediction[0])
confidence= prediction[0][idxs] * 100

# Fetching the Predicted label and confidence score and put it on image 
font_size = 1
text = "Label: {}, {:.2f}%".format(CLASSES[idxs], confidence)
cv2.putText(image, text, (20, 40),  cv2.FONT_HERSHEY_COMPLEX,font_size, (100, 20, 255), 4)

# Diplay the image 
plt.imshow(image[:,:,::-1])
plt.axis("off")
````

#### <font style="color:rgb(134,19,348)"> Image Classification DNN Pipeline in Chunks </font>
Above we have seen whole process of claassification with pretty much detail.for onwards lets create few of chunks of this whole process so whenever we want to classify an image we won't use or do the whole calculation istead just calling a funciton and get done our task   

### <font style="color:rgb(134,19,348)"> Initialize the Method </font>

````python
def init_classify(usegpu=False):
    global net,classes
    rows = open('Model/synset_words.txt').read().strip().split("\n")
    classes = [r[r.find(" ") + 1:].split(",")[0] for r in rows]
    weight='Model/bvlc_googlenet.caffemodel'
    arch='Model/bvlc_googlenet.prototxt'
    net = cv2.dnn.readNetFromCaffe(arch, weight)   
    if usegpu:
        net.setPreferableTarget( cv2.dnn.DNN_TARGET_OPENCL)
````

### <font style="color:rgb(134,19,348)">  Main Method </font>

````python
def classify(image=[],usercam=False,n=1,returndata=False,size=0.7):
    from tkinter import filedialog
    from tkinter import Tk
    import cv2
    import numpy as np

    if np.size(image) == 0:

        if usercam:
            cap=cv2.VideoCapture(0)
            ret,frame=cap.read()
            image= cv2.flip(frame,1)
            cap.release()
        else:
            root = Tk()
            root.filename =  filedialog.askopenfilename(title = "Select Image")
            root.withdraw()
            image = cv2.imread(root.filename)

            
    blob = cv2.dnn.blobFromImage(cv2.resize(image, (224, 224),(104, 117, 123)))
    net.setInput(blob)
    prediction = net.forward()
    
    idxs = np.argsort(prediction[0])[::-1][:n]
    class_data=[]
    Height = 400
    scale = Height / image.shape[0]
    image = cv2.resize(image, None, fx=scale, fy=scale)
    
    for (i, idx) in enumerate(idxs):
        confidence= prediction[0][idx] * 100
        text = "Label: {}, {:.2f}%".format(classes[idx], confidence)
        cv2.putText(image, text, (5, int(size*24) + i*int(size*30)),  cv2.FONT_HERSHEY_COMPLEX,size, (100, 20, 255), 2)
        class_data.append((classes[idx],confidence))
        

    if  returndata:
        return image,class_data
 
    else:
        cv2.imshow("Image", image)
        cv2.waitKey(0)
        cv2.destroyAllWindows() 
````

####  <font style="color:rgb(134,19,348)"> Calling the initlization Method </font>

````python
init_classify()
````

#### <font style="color:rgb(134,19,348)"> Calling Classify with default args </font>

````python
classify()
````

````python
image,result= classify(n=3,returndata=True)
for res in result:
   print('Label {} Confidence: {:.2f}%'.format(res[0],res[1]))
````

## <font style="color:rgb(134,19,348)"> Image Classification On Video </font>

````python
from tkinter import filedialog
from tkinter import Tk
import cv2
import numpy as np
import time

fps=0
init_classify()  #use gpu to increase speed
cap=cv2.VideoCapture(0) 
while(True):    
    
    start_time = time.time()
    ret,frame=cap.read() 
    image=cv2.flip(frame,1) 
    image,_= classify(image,n=3,returndata=True)
    cv2.putText(image, 'FPS: {:.2f}'.format(fps), (370, 20), cv2.FONT_HERSHEY_COMPLEX,0.8, (255, 20, 55), 2)
    cv2.imshow("Image",image)
    fps= (1.0 / (time.time() - start_time))
    k = cv2.waitKey(1)
    x = cv2.getWindowProperty("Image",0)
    if k == 27 or x < 0:
        break
        
cap.release() 
cv2.destroyAllWindows() 
````
